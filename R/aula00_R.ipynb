```r
# Exemplo de código R no Colab

print('Olá, Mundo! Este é R rodando no Colab.')

# Cria dois vetores simples
x <- 1:10
y <- x^2

# Gera um gráfico de dispersão com R base
plot(x, y, main = 'Gráfico Simples em R (Base)', xlab = 'Eixo X', ylab = 'Eixo Y')

# Exemplo de instalação e uso de pacote R (agora ativado)
# Verifica se o pacote ggplot2 já está instalado e instala se não estiver
if (!requireNamespace("ggplot2", quietly = TRUE)) {
  install.packages("ggplot2")
}
library(ggplot2)             # Carrega o pacote ggplot2

df_plot <- data.frame(x = x, y = y)

# Gera um gráfico de dispersão com ggplot2
ggplot(df_plot, aes(x, y)) +
  geom_point() +
  labs(title='Gráfico com ggplot2', x = 'Eixo X', y = 'Eixo Y') +
  theme_minimal()

# Adicionando um exemplo mais complexo com ggplot2
set.seed(123) # para reprodutibilidade
df_complex <- data.frame(
  categoria = sample(c('A', 'B', 'C'), 50, replace = TRUE),
  valor_x = runif(50, 1, 10),
  valor_y = rnorm(50, 5, 2)
)

# Gráfico de dispersão com cores por categoria e linha de tendência
ggplot(df_complex, aes(x = valor_x, y = valor_y, color = categoria)) +
  geom_point(alpha = 0.7) +
  geom_smooth(method = "lm", se = FALSE) +
  facet_wrap(~ categoria) +
  labs(title = 'Gráfico Mais Complexo com ggplot2',
       x = 'Valor X',
       y = 'Valor Y') +
  theme_bw()
```

## 2. Instalação Local do R

Para um desenvolvimento R mais robusto e completo, é altamente recomendável instalar o R e o RStudio em sua máquina local. Isso proporciona um ambiente de trabalho mais estável e com mais recursos do que o Colab para projetos R maiores.

### Passos para Instalar R (o software estatístico base):

1.  **Visite o CRAN (Comprehensive R Archive Network):** Acesse o site oficial do R para downloads: [https://cran.r-project.org/](https://cran.r-project.org/)
2.  **Selecione seu Sistema Operacional:**
    *   **Windows:** Clique em "Download R for Windows" e, em seguida, em "install R for the first time". Baixe o arquivo executável (`.exe`) e execute-o.
    *   **macOS:** Clique em "Download R for (Mac) OS X". Baixe o arquivo do instalador (`.pkg`) mais recente e execute-o.
    *   **Linux:** Siga as instruções específicas para sua distribuição Linux, que geralmente envolvem adicionar o repositório CRAN ao seu sistema e usar o gerenciador de pacotes (por exemplo, `sudo apt-get install r-base` para distribuições baseadas em Debian/Ubuntu).
3.  **Siga as Instruções do Instalador:** Execute o instalador e siga as instruções padrão. Geralmente, as configurações padrão são adequadas, a menos que você tenha necessidades específicas.

## 3. Instalação Local do RStudio

RStudio é um Ambiente de Desenvolvimento Integrado (IDE) amplamente utilizado e altamente recomendado para R. Ele aprimora significativamente a experiência de programação em R, oferecendo diversas ferramentas em uma única interface.

### Passos para Instalar o RStudio Desktop:

1.  **Visite o Site do RStudio:** Acesse a página de download do RStudio Desktop: [https://posit.co/download/rstudio-desktop/](https://posit.co/download/rstudio-desktop/)
2.  **Escolha a Versão Gratuita (Open Source Edition):** Role a página para baixo e localize a seção "RStudio Desktop Open Source Edition". Clique no botão de download para esta versão, que é gratuita e suficiente para a maioria dos usuários.
3.  **Baixe o Instalador para Seu Sistema Operacional:** O site geralmente detecta seu sistema operacional automaticamente e sugere o instalador apropriado (Windows, macOS ou Linux). Baixe o arquivo correspondente.
4.  **Siga as Instruções do Instalador:** Execute o instalador e siga as instruções padrão. O RStudio geralmente detecta automaticamente sua instalação do R base.

### Por que usar o RStudio?

O RStudio oferece uma interface poderosa com múltiplos painéis e funcionalidades, incluindo:
*   **Editor de Código:** Com destaque de sintaxe, autocompletar e recursos de depuração.
*   **Console Interativo:** Para executar comandos R diretamente e ver os resultados.
*   **Painéis de Visualização:** Para ver o ambiente (variáveis, funções), histórico de comandos, arquivos, gráficos e pacotes instalados.
*   **Gerenciamento de Projetos:** Ajuda a organizar seus arquivos e códigos.
*   **Integração com Controle de Versão:** Suporte nativo para Git.
*   **Criação de Relatórios:** Facilita a criação de documentos dinâmicos com R Markdown.

Após a instalação, abra o RStudio para começar a programar em R em um ambiente completo e produtivo!
