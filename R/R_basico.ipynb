## Funções em R

Funções são blocos de código reutilizáveis. R já possui muitas funções embutidas, mas também podemos criar as nossas. Abaixo, um exemplo de função para calcular o conteúdo GC de uma sequência de DNA.

```r
# Criando sua própria função para calcular o conteúdo GC
calculate_gc_content <- function(sequence) {
  sequence <- toupper(sequence)
  gc_count <- sum(gregexpr("G", sequence)[[1]] > 0) + sum(gregexpr("C", sequence)[[1]] > 0)
  total_length <- nchar(sequence)
  if (total_length == 0) {
    return(0)
  } else {
    return((gc_count / total_length) * 100)
  }
}

# Testando a função
seq1 <- "ATGCGTACGTAGCTAGCATG"
gc_seq1 <- calculate_gc_content(seq1)
print(paste("Sequência 1:", seq1, ", Conteúdo GC:", round(gc_seq1, 2), "%"))
```

## Estruturas de Dados Avançadas: Data Frames

Data Frames são essenciais para análise de dados tabulares em R, funcionando como tabelas. Cada coluna pode ter um tipo de dado diferente.

```r
# Exemplo: Dados de expressão gênica para vários genes e amostras
gene_ids <- c("GeneA", "GeneB", "GeneC")
expression_sample1 <- c(10.2, 5.8, 11.5)

genome_data <- data.frame(
  GeneID = gene_ids,
  Sample1_Expression = expression_sample1
)

print("Data Frame de Dados Genômicos (Exemplo):")
print(genome_data)
```
