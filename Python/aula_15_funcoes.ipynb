# Python: Funções e Pacotes

À medida que seus scripts de bioinformática crescem em complexidade, a organização do código torna-se crucial. **Funções** permitem encapsular blocos de código reutilizáveis, enquanto **pacotes** (ou bibliotecas/módulos) fornecem coleções de funções e ferramentas pré-construídas por outros. Juntos, eles são a espinha dorsal da programação eficiente e modular em Python para aplicações biológicas.

## 1. Funções em Python

Uma função é um bloco de código nomeado e reutilizável que executa uma tarefa específica. Elas ajudam a organizar o código, torná-lo mais legível e evitar a repetição.

### Definindo e Chamando Funções

Funções são definidas usando a palavra-chave `def`, seguida do nome da função, parênteses para os parâmetros (se houver) e dois pontos. O corpo da função é indentado. A palavra-chave `return` é usada para enviar um valor de volta após a execução da função.

```python
# Exemplo Biológico 1: Calcular o Conteúdo GC de uma Sequência de DNA
def calcular_gc_content(sequencia_dna):
    """Calcula a porcentagem de conteúdo GC em uma sequência de DNA."""
    contagem_g = sequencia_dna.upper().count('G')
    contagem_c = sequencia_dna.upper().count('C')
    comprimento_total = len(sequencia_dna)

    if comprimento_total == 0:
        return 0
    else:
        return ((contagem_g + contagem_c) / comprimento_total) * 100

# Chamando a função
minha_sequencia = 'ATGCGTACGTAGCTAGCATGC'
pc_gc = calcular_gc_content(minha_sequencia)
print(f"A sequência '{minha_sequencia}' tem {pc_gc:.2f}% de conteúdo GC.")

sequencia_vazia = ''
pc_gc_vazia = calcular_gc_content(sequencia_vazia)
print(f"A sequência '{sequencia_vazia}' tem {pc_gc_vazia:.2f}% de conteúdo GC.")
```

### Funções com Múltiplos Parâmetros e Retorno

Funções podem aceitar múltiplos argumentos e retornar múltiplos valores (como uma tupla).

```python
# Exemplo Biológico 2: Transcrever DNA para RNA
def transcrever_dna_para_rna(sequencia_dna):
    """Transcreve uma sequência de DNA (sentido) para RNA, substituindo T por U."""
    # Converte para maiúsculas para garantir consistência e substitui T por U
    rna_sequencia = sequencia_dna.upper().replace('T', 'U')
    return rna_sequencia

# Chamando a função
sequencia_1 = 'ATGCGTACGTAGCTAGCATGC'
rna_sequencia_1 = transcrever_dna_para_rna(sequencia_1)
print(f"DNA: {sequencia_1}\nRNA: {rna_sequencia_1}")

# Exemplo Biológico 3: Calcular Peso Molecular de uma Proteína
def calcular_peso_molecular_proteina(sequencia_proteina, pesos_aminoacidos):
    """Calcula o peso molecular de uma proteína dada sua sequência e um dicionário de pesos de AA."""
    peso_total = 0
    for aminoacido in sequencia_proteina.upper():
        peso_total += pesos_aminoacidos.get(aminoacido, 0) # Usa .get para evitar erros com AA desconhecidos
    return peso_total

# Dicionário de pesos moleculares médios para aminoácidos (exemplo)
pesos_aa = {
    'A': 89.09,
    'C': 121.16,
    'D': 133.10,
    'E': 147.13,
    'F': 165.19,
    'G': 75.07,
    'H': 155.16,
    'I': 131.17,
    'K': 146.19,
    'L': 131.17,
    'M': 149.21,
    'N': 132.12,
    'P': 115.13,
    'Q': 146.15,
    'R': 174.20,
    'S': 105.09,
    'T': 119.12,
    'V': 117.15,
    'W': 204.23,
    'Y': 181.19
}

proteina_exemplo = 'MVLSPADKTNVKAAWGKV'
peso_molecular = calcular_peso_molecular_proteina(proteina_exemplo, pesos_aa)
print(f"\nA proteína '{proteina_exemplo}' tem um peso molecular de {peso_molecular:.2f} Da.")
```

## 2. Pacotes e Módulos em Python

Um **módulo** é um arquivo Python (`.py`) que contém funções, classes e variáveis. Um **pacote** é uma coleção de módulos organizados em diretórios. Eles permitem reutilizar código de forma mais ampla e acessar ferramentas poderosas desenvolvidas por outros.

### Importando Módulos e Pacotes

Usamos a palavra-chave `import` para trazer módulos ou partes de módulos para o nosso script.

```python
# Importando um módulo padrão: math
import math

# Exemplo: Arredondar um valor de concentração para cima
concentracao_dna = 50.75 # ng/uL
concentracao_arredondada = math.ceil(concentracao_dna) # Arredonda para o inteiro mais próximo acima
print(f"Concentração de DNA original: {concentracao_dna}, arredondada para cima: {concentracao_arredondada}")

# Importando partes específicas de um módulo
from math import floor, pi

# Exemplo: Arredondar para baixo e usar o valor de pi
valor_ph = 6.999999999999999
ph_arredondado = floor(valor_ph)
print(f"pH original: {valor_ph}, arredondado para baixo: {ph_arredondado}")
print(f"O valor de Pi é: {pi:.4f}")
```

```python
# imprime salários formatados
def calcular_salarios(valores):
  for valor in valores:
    print(f"R${valor:.2f}")

lista_salarios = [1800, 2000, 3000, 4000, 5000]
calcular_salarios(lista_salarios)
```

### Utilizando Bibliotecas Externas (Ex: Biopython)

Para bioinformática, existem bibliotecas especializadas como `Biopython`, que oferece objetos e funções para manipulação de sequências, acesso a bancos de dados biológicos, análise filogenética, etc.

Para usar uma biblioteca externa, você geralmente precisa instalá-la primeiro (no Colab, isso é feito com `!pip install`).

**Observação**: Não vamos instalar o Biopython aqui para manter o foco na introdução a funções e pacotes, mas este é um exemplo de como você o usaria.

```python
# Exemplo Conceitual: Usando Biopython para manipulação de sequências
# Imagine que você já instalou o Biopython com '!pip install biopython'

from Bio.Seq import Seq
import Bio.SeqUtils

min_seq_dna_biopython = Seq("ATGCGTACGTAGCTAGCATGC")
print(f"Sequência Biopython: {min_seq_dna_biopython}")

gc_content_fraction = Bio.SeqUtils.gc_fraction(min_seq_dna_biopython)
print(f"Conteúdo GC: {gc_content_fraction * 100:.2f}%")
```

```python
# Exemplo de tradução de uma sequência de DNA usando Biopython
from Bio.Seq import Seq

# Sequência de DNA de exemplo (deve ser uma região codificante)
dna_sequence = Seq("ATGCATGCATGCATGCATGCATGCATGCATGCATGCATGC")

# Transcreve DNA para RNA
rna_sequence = dna_sequence.transcribe()

# Traduz RNA para proteína
protein_sequence = rna_sequence.translate()

print(f"Sequência de DNA: {dna_sequence}")
print(f"Sequência de RNA (transcrita): {rna_sequence}")
print(f"Sequência de Proteína (traduzida): {protein_sequence}")

# Você também pode traduzir diretamente do DNA, especificando a tabela de código genético (padrão é 1, Standard Code)
protein_from_dna = dna_sequence.translate(table=1)
print(f"Sequência de Proteína (traduzida diretamente do DNA): {protein_from_dna}")
```

## Conclusão

Funções e pacotes são pilares da programação em Python, especialmente em campos como a biologia computacional. Funções permitem criar código modular e reutilizável para tarefas específicas (como calcular GC content ou traduzir sequências). Pacotes fornecem um ecossistema rico de ferramentas que aceleram o desenvolvimento, permitindo que você se baseie no trabalho de outros (como a manipulação avançada de sequências do Biopython). Dominar esses conceitos é fundamental para escrever código eficiente, limpo e escalável em suas análises biológicas.
