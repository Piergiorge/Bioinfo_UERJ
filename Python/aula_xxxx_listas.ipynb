# Python: Listas

No vasto mundo da biologia e bioinformática, frequentemente nos deparamos com a necessidade de organizar e manipular coleções de dados. Seja uma série de sequências genéticas, uma lista de genes de interesse, os resultados de múltiplas amostras experimentais ou as coordenadas de um genoma, as **listas** em Python são uma ferramenta fundamental para gerenciar essas informações de forma ordenada e flexível.

## 1. O que são Listas?

Uma lista é uma coleção ordenada e mutável de itens. Ao contrário das strings (que armazenam apenas caracteres), as listas podem conter itens de diferentes tipos de dados (inteiros, floats, strings, e até outras listas!). São definidas por colchetes `[]`, com os itens separados por vírgulas.

## 2. Criando Listas

As listas podem ser criadas de diversas formas, seja com itens predefinidos ou a partir de outras estruturas de dados.

```python
# Exemplo 1: Lista de Genes de Interesse
genes_interesse = ['BRCA1', 'TP53', 'ATM', 'KRAS', 'EGFR']
print(f"Genes de interesse: {genes_interesse}")

# Exemplo 2: Lista de Níveis de Expressão Gênica (mistura de tipos)
valores_expressao = [10.5, 23.1, 15.2, 5.8, 'baixo', 30.0]
print(f"Valores de expressão: {valores_expressao}")

# Exemplo 3: Lista de Coordenadas Genômicas (tuplas dentro de uma lista)
coordenadas_brca1 = ('chr17', 43044295, 43125483)
coordenadas_tp53 = ('chr17', 7661779, 7687550)
regioes_genomicas = [coordenadas_brca1, coordenadas_tp53]
print(f"Regiões genômicas: {regioes_genomicas}")

# Exemplo 4: Lista vazia
resultados_experimento = []
print(f"Lista vazia para resultados: {resultados_experimento}")
```

## 3. Acessando Elementos

Elementos em uma lista são acessados usando índices (começando do `0`). Você também pode usar indexação negativa (o último item é `-1`) e fatiamento (slicing) para obter sub-listas.

```python
genes_interesse = ['BRCA1', 'TP53', 'ATM', 'KRAS', 'EGFR']

# Acessando o primeiro gene
primeiro_gene = genes_interesse[0]
print(f"Primeiro gene: {primeiro_gene}")

# Acessando o último gene
ultimo_gene = genes_interesse[-1]
print(f"Último gene: {ultimo_gene}")

# Acessando um intervalo de genes (fatiamento)
遺伝子群 = genes_interesse[1:4] # Do índice 1 (incluído) ao 4 (excluído)
print(f"Genes do índice 1 ao 3: {遺伝子群}")

# Acessando do início até um ponto
primeiros_dois = genes_interesse[:2]
print(f"Primeiros dois genes: {primeiros_dois}")

# Acessando de um ponto até o final
restantes = genes_interesse[2:]
print(f"Genes a partir do índice 2: {restantes}")
```

## 4. Modificando Listas

Listas são mutáveis, o que significa que seus elementos podem ser alterados, adicionados ou removidos após a criação da lista. Isso é crucial para dados biológicos que podem ser atualizados ou corrigidos.

```python
genes_amostra = ['GENEA', 'GENEB', 'GENEC']
print(f"Lista original: {genes_amostra}")

# Alterar um elemento
genes_amostra[1] = 'GENEX'
print(f"Após alterar 'GENEB' para 'GENEX': {genes_amostra}")

# Adicionar um elemento ao final (append)
genes_amostra.append('GENED')
print(f"Após adicionar 'GENED' com append: {genes_amostra}")

# Inserir um elemento em uma posição específica (insert)
genes_amostra.insert(1, 'GENEZ') # Insere 'GENEZ' no índice 1
print(f"Após inserir 'GENEZ' no índice 1: {genes_amostra}")

# Estender a lista com outra lista (extend)
genes_novos = ['GENEF', 'GENEG']
genes_amostra.extend(genes_novos)
print(f"Após estender com 'GENEF', 'GENEG': {genes_amostra}")

# Remover um elemento pelo valor (remove)
genes_amostra.remove('GENEX')
print(f"Após remover 'GENEX': {genes_amostra}")

# Remover um elemento pelo índice e retorná-lo (pop)
gene_removido = genes_amostra.pop(0) # Remove o primeiro elemento
print(f"Após remover o primeiro gene ('{gene_removido}') com pop: {genes_amostra}")

# Remover um elemento pelo índice (del)
del genes_amostra[-1] # Remove o último elemento
print(f"Após remover o último gene com del: {genes_amostra}")

# Limpar todos os elementos da lista
genes_amostra.clear()
print(f"Após limpar a lista: {genes_amostra}")
```

## 5. Operações Comuns com Listas

Python oferece operadores que funcionam de forma intuitiva com listas.

```python
# Concatenação de listas (união de resultados)
resultados_estudo1 = ['Mutação A', 'SNP B']
resultados_estudo2 = ['Mutação C', 'SNP D', 'Mutação A']
todos_resultados = resultados_estudo1 + resultados_estudo2
print(f"Todos os resultados: {todos_resultados}")

# Repetição de listas (preparar placas de poços)
volume_padrao = [100, 200]
placa_96_poços = volume_padrao * 48 # Repete 48 vezes para 96 poços
# print(f"Placa de 96 poços (apenas o início): {placa_96_poços[:5]}...")
print(f"Tamanho da placa de 96 poços: {len(placa_96_poços)}")

# Comprimento da lista (len)
num_resultados = len(todos_resultados)
print(f"Número total de resultados: {num_resultados}")

# Verificação de associação (in/not in)
gene_procurado = 'TP53'
genes_expressos = ['BRCA1', 'TP53', 'CDKN2A']
if gene_procurado in genes_expressos:
    print(f"'{gene_procurado}' está na lista de genes expressos.")
else:
    print(f"'{gene_procurado}' NÃO está na lista de genes expressos.")

# Min, Max, Sum (para dados numéricos)
dados_quantitativos = [12.5, 8.2, 20.1, 5.0, 18.7]
print(f"Mínimo: {min(dados_quantitativos)}")
print(f"Máximo: {max(dados_quantitativos)}")
print(f"Soma: {sum(dados_quantitativos)}")
```

## 6. Métodos Úteis de Listas

Além das operações, as listas possuem vários métodos embutidos para manipulação mais específica.

```python
temperaturas_reacao = [25, 37, 55, 60, 30]

# Ordenar a lista (sort)
temperaturas_reacao.sort()
print(f"Temperaturas ordenadas: {temperaturas_reacao}")

# Inverter a ordem (reverse)
temperaturas_reacao.reverse()
print(f"Temperaturas invertidas: {temperaturas_reacao}")

# Contar ocorrências de um item (count)
dupl_amostras = ['A', 'B', 'A', 'C', 'B', 'A']
contagem_A = dupl_amostras.count('A')
print(f"'A' aparece {contagem_A} vezes.")

# Encontrar o índice de um item (index)
indice_C = dupl_amostras.index('C')
print(f"'C' está no índice: {indice_C}")

# Copiar uma lista
# CUIDADO: Atribuição simples copia a REFERÊNCIA, não a lista!
lista_original = ['GeneX', 'GeneY']
lista_referencia = lista_original
lista_copia = lista_original.copy()

lista_original.append('GeneZ')
print(f"Original: {lista_original}") # GeneZ está aqui
print(f"Referência: {lista_referencia}") # GeneZ TAMBÉM está aqui!
print(f"Cópia: {lista_copia}") # GeneZ NÃO está aqui (cópia independente)
```

## 7. Iterando sobre Listas

A iteração é a forma mais comum de processar cada item de uma lista, como analisar cada gene ou cada resultado de uma amostra.

```python
sequencia_peptidica = ['M', 'V', 'L', 'S', 'P', 'A', 'D', 'K']

print("Aminoácidos na sequência:")
for aminoacido in sequencia_peptidica:
    print(aminoacido)

# Iterar com índice (se precisar da posição)
print("\nAminoácidos com seus índices:")
for i, aminoacido in enumerate(sequencia_peptidica):
    print(f"Índice {i}: {aminoacido}")
```

## 8. List Comprehensions

List comprehensions oferecem uma forma concisa e eficiente de criar novas listas a partir de listas existentes. São muito úteis para filtragem e transformação de dados biológicos.

```python
genes_encontrados = ['BRCA1', 'TP53', 'CDKN2A', 'APC', 'KRAS', 'NF1']

# Exemplo 1: Filtrar genes que começam com 'B'
genes_com_b = [gene for gene in genes_encontrados if gene.startswith('B')]
print(f"Genes que começam com 'B': {genes_com_b}")

# Exemplo 2: Converter nomes de genes para minúsculas
genes_minusculas = [gene.lower() for gene in genes_encontrados]
print(f"Genes em minúsculas: {genes_minusculas}")

# Exemplo 3: Selecionar apenas genes com mais de 4 letras
genes_longos = [gene for gene in genes_encontrados if len(gene) > 4]
print(f"Genes com mais de 4 letras: {genes_longos}")

# Exemplo 4: Aplicar uma função a cada elemento (simular nível de expressão)
niveis_expressao_base = [10, 25, 5, 18, 30]
niveis_expressao_dobrados = [x * 2 for x in niveis_expressao_base]
print(f"Níveis de expressão dobrados: {niveis_expressao_dobrados}")
```
