## O que é o Google Colab?

Google Colaboratory (ou Colab) é um ambiente de notebook Jupyter gratuito que é executado na nuvem do Google. Ele permite que você escreva e execute código Python (ou R) diretamente no seu navegador, sem a necessidade de nenhuma configuração.

**Principais recursos:**
*   **Ambiente Gratuito:** Acesso a GPUs e TPUs gratuitamente.
*   **Compartilhamento Fácil:** Compartilhe seus notebooks com qualquer pessoa.
*   **Integração:** Conectado ao seu Google Drive para salvar e carregar notebooks e dados.
*   **Pré-instalado:** Muitas bibliotecas populares de ciência de dados já vêm pré-instaladas.

## Como Rodar Python Localmente

Embora o Colab seja excelente, muitas vezes você precisará rodar Python em sua própria máquina. Aqui estão os passos básicos:

1.  **Instalar Python:** Baixe e instale a versão mais recente do Python em [python.org](https://www.python.org/downloads/). Para iniciantes, uma distribuição como [Anaconda](https://www.anaconda.com/products/individual) ou [Miniconda](https://docs.conda.io/en/latest/miniconda.html) é altamente recomendada, pois inclui Python e muitas bibliotecas científicas populares.

2.  **Ambientes Virtuais:** É uma boa prática usar ambientes virtuais para gerenciar dependências de projetos. Isso evita conflitos entre as bibliotecas de diferentes projetos.
    *   Com `venv` (módulo integrado do Python):
        ```bash
        python -m venv meu_ambiente
        source meu_ambiente/bin/activate  # No Windows: meu_ambiente\Scripts\activate
        ```
    *   Com `conda` (se você usa Anaconda/Miniconda):
        ```bash
        conda create --name meu_ambiente python=3.9
        conda activate meu_ambiente
        ```

3.  **Instalar Bibliotecas:** Dentro do seu ambiente virtual, use `pip` para instalar as bibliotecas necessárias:
    ```bash
    pip install pandas numpy matplotlib
    ```

4.  **Escrever e Executar seu Código:**
    *   Crie um arquivo Python (ex: `meu_script.py`) em um editor de texto ou IDE (como VS Code, PyCharm).
    *   Salve seu código Python nesse arquivo.
    *   Abra seu terminal/prompt de comando, navegue até a pasta onde salvou o arquivo e execute-o:
        ```bash
        python meu_script.py
        ```
