# Python: Tipo Booleano (`bool`), Tabelas Verdade e Condicionais (`if`/`elif`/`else`)

Neste notebook, vamos explorar o tipo booleano, que é essencial para tomar decisões em nossos programas. Veremos como os operadores lógicos funcionam através das tabelas verdade e como usar as estruturas condicionais `if`, `elif` e `else` para executar blocos de código com base em condições booleanas.

## 1. Tipo Booleano (`bool`)

O tipo `bool` representa valores de verdade e falsidade. Em Python, existem duas constantes booleanas:
- `True`: representa verdadeiro.
- `False`: representa falso.

Ambas as constantes são escritas com a primeira letra maiúscula.

```python
# Exemplos de valores booleanos

valor_verdadeiro = True
valor_falso = False

print(f"Valor verdadeiro: {valor_verdadeiro}, Tipo: {type(valor_verdadeiro)}")
print(f"Valor falso: {valor_falso}, Tipo: {type(valor_falso)}")

# Expressões que resultam em booleanos
print(f"10 > 5: {10 > 5}")
print(f"10 == 5: {10 == 5}")
print(f"'Python' == 'python': {'Python' == 'python'}")
```

## 2. Operadores Lógicos (Tabelas Verdade)

Os operadores lógicos `and`, `or` e `not` são usados para combinar e manipular valores booleanos. As tabelas verdade mostram o resultado de todas as combinações possíveis.

### 2.1. Operador `and` (E Lógico)

O operador `and` retorna `True` se *ambas* as condições forem verdadeiras. Caso contrário, retorna `False`.

**Tabela Verdade para `and`:**

| A     | B     | A and B |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |

```python
# Exemplos com 'and'

print(f"True and True: {True and True}")
print(f"True and False: {True and False}")
print(f"False and True: {False and True}")
print(f"False and False: {False and False}")

idade = 25
tem_carteira = True

if idade >= 18 and tem_carteira:
    print("Pode dirigir.")
else:
    print("Não pode dirigir.")
```

### 2.2. Operador `or` (OU Lógico)

O operador `or` retorna `True` se *pelo menos uma* das condições for verdadeira. Retorna `False` apenas se ambas as condições forem falsas.

**Tabela Verdade para `or`:**

| A     | B     | A or B  |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | True    |
| False | True  | True    |
| False | False | False   |

```python
# Exemplos com 'or'

print(f"True or True: {True or True}")
print(f"True or False: {True or False}")
print(f"False or True: {False or True}")
print(f"False or False: {False or False}")

tem_passaporte = True
tem_visto = False

if tem_passaporte or tem_visto:
    print("Pode viajar internacionalmente (se um dos dois for suficiente).")
else:
    print("Não pode viajar internacionalmente.")
```

### 2.3. Operador `not` (NÃO Lógico)

O operador `not` inverte o valor booleano da condição. Se for `True`, torna-se `False`, e vice-versa.

**Tabela Verdade para `not`:**

| A     | not A |
|-------|-------|
| True  | False |
| False | True  |

```python
# Exemplos com 'not'

print(f"not True: {not True}")
print(f"not False: {not False}")

usuario_logado = False

if not usuario_logado:
    print("Por favor, faça login.")
else:
    print("Bem-vindo!")
```

## 3. Estruturas Condicionais: `if`, `elif`, `else`

As estruturas condicionais permitem que você execute diferentes blocos de código com base na avaliação de condições booleanas. Isso é essencial para criar programas que respondam a diferentes cenários.

### 3.1. `if` Statement

O bloco `if` executa um conjunto de instruções **se** a condição booleana for `True`.

```python
# Exemplo de 'if'

temperatura = 28

if temperatura > 25:
    print("Está calor! \ud83c\udf1e")
```

### 3.2. `if`...`else` Statement

O bloco `if`...`else` executa um conjunto de instruções **se** a condição for `True`, e um bloco alternativo de instruções **se** a condição for `False`.

```python
# Exemplo de 'if'...'else'

idade_usuario = 17

if idade_usuario >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

### 3.3. `if`...`elif`...`else` Statement

Use `elif` (abreviação de 'else if') para verificar múltiplas condições. O Python avalia as condições em ordem e executa o bloco de código da primeira condição que for `True`. Se nenhuma condição `if` ou `elif` for `True`, o bloco `else` (se existir) será executado.

```python
# Exemplo de 'if'...'elif'...'else'

nota = 75

if nota >= 90:
    print("Conceito: A")
elif nota >= 80:
    print("Conceito: B")
elif nota >= 70:
    print("Conceito: C")
elif nota >= 60:
    print("Conceito: D")
else:
    print("Conceito: F")

# Outro exemplo com strings
cor_semaforo = "amarelo"

if cor_semaforo == "verde":
    print("Siga!")
elif cor_semaforo == "amarelo":
    print("Atenção!")
else:
    print("Pare!")
```

### 3.4. Aninhamento de Condicionais

Você pode aninhar estruturas condicionais dentro de outras para criar lógicas mais complexas.

```python
# Exemplo de condicionais aninhadas

possui_ingresso = True
tem_pipoca = False

if possui_ingresso:
    print("Pode entrar no cinema.")
    if tem_pipoca:
        print("Aproveite seu filme com pipoca!")
    else:
        print("Que tal comprar uma pipoca?")
else:
    print("Precisa de um ingresso para entrar.")
```

Dominar esses conceitos é fundamental para escrever programas flexíveis e inteligentes que podem responder a diversas entradas e situações.
