## 2. Operadores Lógicos em Biologia (Tabelas Verdade)

Os operadores lógicos `and`, `or` e `not` são usados para combinar e manipular condições biológicas. As tabelas verdade nos ajudam a entender o resultado de diferentes combinações.

## 1. Tipo Booleano (`bool`): Verdadeiro ou Falso na Biologia

O tipo `bool` representa valores de verdade e falsidade. Na biologia, podemos pensar em características ou estados que são claramente 'verdadeiros' ou 'falsos'. Por exemplo, um organismo 'é vertebrado' (True) ou 'não é vertebrado' (False). Em Python, existem duas constantes booleanas:
- `True`: representa verdadeiro.
- `False`: representa falso.

Ambas as constantes são escritas com a primeira letra maiúscula.

```python
# Exemplos de valores booleanos em biologia

organismo_vertebrado = True
tem_clorofila = False

print(f"O organismo é vertebrado: {organismo_vertebrado}, Tipo: {type(organismo_vertebrado)}")
print(f"O organismo tem clorofila: {tem_clorofila}, Tipo: {type(tem_clorofila)}")

# Expressões que resultam em booleanos para dados biológicos
quantidade_oxigenio = 21 # em %
temperatura_ambiente = 25 # em Celsius

print(f"Oxigênio > 20%: {quantidade_oxigenio > 20}") # Nível de oxigênio ideal para muitas espécies
print(f"Temperatura == 25: {temperatura_ambiente == 25}") # Temperatura ótima para algumas enzimas
print(f"'Bactéria' == 'vírus': {'Bactéria' == 'vírus'}") # Comparação de tipos de microrganismos
```

## 2. Operadores Lógicos (Tabelas Verdade)

Os operadores lógicos `and`, `or` e `not` são usados para combinar e manipular valores booleanos. As tabelas verdade mostram o resultado de todas as combinações possíveis.

### 2.1. Operador `and` (E Lógico): Ambas as Condições Biológicas Verdadeiras

O operador `and` retorna `True` se *ambas* as condições biológicas forem verdadeiras. Caso contrário, retorna `False`.

**Tabela Verdade para `and`:**

| A     | B     | A and B |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |

```python
# Exemplos com 'and' em biologia

# Condições para ser um mamífero aquático
tem_glandulas_mamarias = True
vive_na_agua = True

print(f"Tem glândulas mamárias E vive na água: {tem_glandulas_mamarias and vive_na_agua}")

# Condições para uma planta realizar fotossíntese eficientemente
tem_luz_suficiente = True
tem_agua_disponivel = False

print(f"Tem luz suficiente E água disponível: {tem_luz_suficiente and tem_agua_disponivel}")

# Exemplo prático: Classificação de um organismo
eh_animal = True
tem_celulas_eucarioticas = True
tem_parede_celular = False

if eh_animal and tem_celulas_eucarioticas and not tem_parede_celular:
    print("Provavelmente é um organismo do Reino Animalia.")
else:
    print("Não se encaixa perfeitamente na descrição de um animal típico.")
```

### 2.2. Operador `or` (OU Lógico): Pelo Menos Uma Condição Biológica Verdadeira

O operador `or` retorna `True` se *pelo menos uma* das condições biológicas for verdadeira. Retorna `False` apenas se ambas as condições forem falsas.

**Tabela Verdade para `or`:**

| A     | B     | A or B  |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | True    |
| False | True  | True    |
| False | False | False   |

```python
# Exemplos com 'or' em biologia

# Condições para ser considerado um organismo voador (ex: pássaro ou inseto)
tem_asas = True
e_leve_o_suficiente = True

print(f"Tem asas OU é leve o suficiente: {tem_asas or e_leve_o_suficiente}")

# Condições para uma espécie ser classificada como invasora (se um dos critérios for atingido)
domina_nicho = False
causa_dano_ecologico = True

if domina_nicho or causa_dano_ecologico:
    print("Esta espécie pode ser considerada invasora.")
else:
    print("Esta espécie não parece ser invasora com base nesses critérios.")
```

### 2.3. Operador `not` (NÃO Lógico): Invertendo o Estado Biológico

O operador `not` inverte o valor booleano da condição. Se for `True`, torna-se `False`, e vice-versa. Por exemplo, `not tem_doenca` significa 'não tem a doença'.

```python
# Exemplos com 'not' em biologia

tem_mutacao_gene_x = False

print(f"Não tem mutação no gene X: {not tem_mutacao_gene_x}")

dados_genoma_analisados = False

if not dados_genoma_analisados:
    print("Os dados do genoma ainda precisam ser analisados.")
else:
    print("Análise do genoma concluída.")
```

## 3. Estruturas Condicionais em Biologia: `if`, `elif`, `else`

As estruturas condicionais permitem que você execute diferentes blocos de código com base na avaliação de condições biológicas. Isso é essencial para criar programas que modelam processos biológicos, classificam organismos ou analisam dados com base em critérios.

### 3.1. `if` Statement: Reagindo a uma Única Condição Biológica

O bloco `if` executa um conjunto de instruções **se** a condição booleana for `True`. Por exemplo, 'se a temperatura estiver acima de um certo limite, um processo biológico é ativado'.

```python
# Exemplo de 'if' em biologia

temperatura_ambiente = 38 # em Celsius

if temperatura_ambiente > 37:
    print(f"A temperatura está acima do ideal para a maioria das enzimas humanas!")
```

### 3.2. `if`...`else` Statement: Duas Possibilidades Biológicas

O bloco `if`...`else` executa um conjunto de instruções **se** a condição biológica for `True`, e um bloco alternativo de instruções **se** a condição for `False`. Pense em 'se o gene X está presente, faça A; senão, faça B'.

```python
# Exemplo de 'if'...'else' em biologia

pH_solucao = 6.5

if pH_solucao < 7.0:
    print("A solução é ácida.")
else:
    print("A solução é neutra ou básica.")
```

### 3.3. `if`...`elif`...`else` Statement: Múltiplas Condições Biológicas

Use `elif` (abreviação de 'else if') para verificar múltiplas condições biológicas em sequência. O Python avalia as condições em ordem e executa o bloco de código da primeira condição que for `True`. Se nenhuma condição `if` ou `elif` for `True`, o bloco `else` (se existir) será executado. Isso é útil para classificação de espécies, análise de diferentes estados de uma doença, etc.

```python
# Exemplo de 'if'...'elif'...'else' em biologia: Classificação de Glicemia

nivel_glicose = 95 # em mg/dL

if nivel_glicose < 70:
    print("Estado: Hipoglicemia")
elif nivel_glicose >= 70 and nivel_glicose <= 99:
    print("Estado: Glicemia Normal")
elif nivel_glicose >= 100 and nivel_glicose <= 125:
    print("Estado: Pré-diabetes")
else:
    print("Estado: Diabetes")

# Outro exemplo: Classificação de tipo de célula
tipo_celula = "eucariótica animal"

if tipo_celula == "eucariótica vegetal":
    print("Célula com parede celular e cloroplastos.")
elif tipo_celula == "eucariótica animal":
    print("Célula sem parede celular, com centríolos.")
elif tipo_celula == "procariótica":
    print("Célula sem núcleo definido e organelas membranosas.")
else:
    print("Tipo de célula não reconhecido.")
```

### 3.4. Aninhamento de Condicionais: Lógicas Biológicas Mais Complexas

Você pode aninhar estruturas condicionais dentro de outras para criar lógicas mais complexas, simulando a hierarquia de decisões ou características biológicas.

```python
# Exemplo de condicionais aninhadas em biologia: Classificação de um animal

e_mamifero = True
produz_leite = True
tem_pelos = False

if e_mamifero:
    print("É um mamífero.")
    if produz_leite:
        print("Este mamífero produz leite para seus filhotes.")
        if tem_pelos:
            print("E também possui pelos.")
        else:
            print("Mas não possui pelos visíveis (ex: golfinho).")
    else:
        print("Parece ser um mamífero atípico que não produz leite (erro na lógica ou característica rara?).")
else:
    print("Não é um mamífero.")
```
