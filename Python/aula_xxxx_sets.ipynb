# Python: Sets em Biologia

Na biologia e bioinformática, muitas vezes lidamos com coleções de itens onde a ordem não importa e onde cada item deve ser único. Pense em uma lista de genes expressos em uma amostra, um conjunto de espécies observadas em um ecossistema, ou mutações genéticas encontradas em um paciente. Para esses cenários, o tipo de dado `set` (conjunto) do Python é uma ferramenta extremamente útil.

## 1. O que são Sets?

Um `set` é uma coleção desordenada de itens únicos. Isso significa:
*   **Desordenado**: Os itens não têm uma ordem definida; você não pode acessá-los por índice.
*   **Único**: Cada item em um conjunto deve ser diferente dos outros. Duplicatas são automaticamente removidas.
*   **Mutável**: Sets podem ter elementos adicionados ou removidos (exceto frozensets).
*   **Itens Imutáveis**: Os elementos de um set devem ser de tipos imutáveis (números, strings, tuplas).

## 2. Criando Sets

```python
# Exemplo 1: Genes Expressos (garantindo unicidade)
genes_amostra_A = {'BRCA1', 'TP53', 'EGFR', 'TP53', 'KRAS'}
print(f"Genes na Amostra A: {genes_amostra_A} (Tipo: {type(genes_amostra_A)})")
print(f"Número de genes únicos: {len(genes_amostra_A)}")

# Exemplo 2: Espécies Encontradas em um Habitat
especies_amazonia = {'Jaguar', 'Arara', 'Capivara', 'Onça-pintada', 'Arara'}
print(f"Espécies na Amazônia: {especies_amazonia}")

# Exemplo 3: Criando um set a partir de uma lista (duplicatas são removidas)
lista_mutacoes = ['deleção', 'substituição', 'inserção', 'deleção', 'inversão']
tipos_mutacao_unicos = set(lista_mutacoes)
print(f"Tipos de mutação únicos: {tipos_mutacao_unicos}")

# Exemplo 4: Set vazio
set_vazio = set()
print(f"Set vazio: {set_vazio} (Tipo: {type(set_vazio)})")
```

## 3. Adicionando e Removendo Elementos

```python
genes_alvo = {'BRCA1', 'TP53'}
print(f"Genes alvo iniciais: {genes_alvo}")

# Adicionar um gene
genes_alvo.add('ATM')
print(f"Após adicionar 'ATM': {genes_alvo}")

# Tentar adicionar um gene que já existe (não fará nada)
genes_alvo.add('TP53')
print(f"Após tentar adicionar 'TP53' novamente: {genes_alvo}")

# Remover um gene
genes_alvo.remove('TP53') # Gerará um erro se o item não existir
print(f"Após remover 'TP53': {genes_alvo}")

# Remover um gene (sem erro se não existir)
genes_alvo.discard('CDKN2A') # Não faz nada se 'CDKN2A' não estiver presente
print(f"Após tentar descartar 'CDKN2A': {genes_alvo}")

# Remover um elemento aleatório (útil em certas situações)
# elemento_removido = genes_alvo.pop()
# print(f"Elemento removido aleatoriamente: {elemento_removido}")
# print(f"Genes alvo restantes: {genes_alvo}")

# Limpar todos os elementos
genes_alvo.clear()
print(f"Após limpar o set: {genes_alvo}")
```

## 4. Operações de Conjunto (Set Operations)

A verdadeira força dos sets está nas suas operações matemáticas de conjunto, que são incrivelmente úteis em análises biológicas.

```python
# Cenário: Análise de Genes em Diferentes Amostras/Vias
genes_via_sinalizacao_A = {'MAPK1', 'ERK1', 'RAS', 'RAF', 'MEK1'}
genes_via_sinalizacao_B = {'AKT', 'RAS', 'RAF', 'PI3K', 'mTOR'}

print(f"\nGenes da Via A: {genes_via_sinalizacao_A}")
print(f"Genes da Via B: {genes_via_sinalizacao_B}")

# União (`union()` ou `|`): Todos os genes presentes em Pelo Menos Uma via
todos_genes_envolvidos = genes_via_sinalizacao_A.union(genes_via_sinalizacao_B)
# Ou: todos_genes_envolvidos = genes_via_sinalizacao_A | genes_via_sinalizacao_B
print(f"\nUnião (Todos os genes envolvidos): {todos_genes_envolvidos}")

# Interseção (`intersection()` ou `&`): Genes Comuns a Ambas as Vias
genes_comuns = genes_via_sinalizacao_A.intersection(genes_via_sinalizacao_B)
# Ou: genes_comuns = genes_via_sinalizacao_A & genes_via_sinalizacao_B
print(f"Interseção (Genes comuns às vias): {genes_comuns}")

# Diferença (`difference()` ou `-`): Genes em A mas Não em B
genes_unicos_A = genes_via_sinalizacao_A.difference(genes_via_sinalizacao_B)
# Ou: genes_unicos_A = genes_via_sinalizacao_A - genes_via_sinalizacao_B
print(f"Diferença (Genes únicos na Via A): {genes_unicos_A}")

# Diferença Simétrica (`symmetric_difference()` ou `^`): Genes em A ou B, mas Não em Ambos
genes_diferentes = genes_via_sinalizacao_A.symmetric_difference(genes_via_sinalizacao_B)
# Ou: genes_diferentes = genes_via_sinalizacao_A ^ genes_via_sinalizacao_B
print(f"Diferença Simétrica (Genes exclusivos de A ou B): {genes_diferentes}")


# Cenário: Mutações Genéticas e Susceptibilidade
mutacoes_paciente = {'BRAF V600E', 'TP53 R175H', 'EGFR L858R'}
mutacoes_susceptibilidade_X = {'BRAF V600E', 'KRAS G12C'}
mutacoes_susceptibilidade_Y = {'TP53 R175H', 'BRCA1 del'}

print(f"\nMutações do Paciente: {mutacoes_paciente}")
print(f"Mutações Susceptibilidade X: {mutacoes_susceptibilidade_X}")
print(f"Mutações Susceptibilidade Y: {mutacoes_susceptibilidade_Y}")

# Subconjunto (`issubset()`): As mutações do paciente são um subconjunto de X?
print(f"As mutações do paciente são um subconjunto da susceptibilidade X? {mutacoes_susceptibilidade_X.issubset(mutacoes_paciente)}")
print(f"As mutações do paciente são um subconjunto da susceptibilidade Y? {mutacoes_susceptibilidade_Y.issubset(mutacoes_paciente)}")

# Superconjunto (`issuperset()`): O paciente tem todas as mutações da susceptibilidade X?
print(f"O paciente tem todas as mutações da susceptibilidade X? {mutacoes_paciente.issuperset(mutacoes_susceptibilidade_X)}")

# Verificando a existência de uma mutação específica
mutacao_especifica = 'EGFR L858R'
if mutacao_especifica in mutacoes_paciente:
    print(f"\nA mutação '{mutacao_especifica}' foi encontrada no paciente.")
else:
    print(f"A mutação '{mutacao_especifica}' não foi encontrada no paciente.")
```

## 5. Frozenset (Sets Imutáveis)

`frozenset` é uma versão imutável de um `set`. Uma vez criado, você não pode adicionar nem remover elementos. Isso os torna hashable, o que significa que podem ser usados como chaves de dicionário ou como elementos de outros sets.

```python
# Exemplo de Frozenset para condições de experimento
condicao_experimento_basal = frozenset({'temperatura_37C', 'pH_7.4', 'meio_DMEM'})
condicao_experimento_estresse = frozenset({'temperatura_37C', 'pH_7.4', 'meio_DMEM', 'estresse_oxidativo'})

print(f"Condição Basal: {condicao_experimento_basal} (Tipo: {type(condicao_experimento_basal)})")
print(f"Condição Estresse: {condicao_experimento_estresse}")

# Frozensets podem ser usados como chaves de dicionário
resultados_experimento = {
    condicao_experimento_basal: 'Expressão normal',
    condicao_experimento_estresse: 'Expressão alterada'
}

print(f"\nResultado para condição basal: {resultados_experimento[condicao_experimento_basal]}")
```

## Conclusão

Sets são uma estrutura de dados poderosa em Python para lidar com coleções de elementos únicos, especialmente quando a ordem não é um fator e você precisa realizar operações de conjunto eficientes. Sua aplicação em biologia é vasta, permitindo análises como a identificação de genes comuns entre vias, mutações compartilhadas entre pacientes, ou a comparação de espécies em diferentes ecossistemas. Dominar os sets é fundamental para otimizar a manipulação e análise de dados biológicos.
