# Python: Strings (Sequências de Caracteres) e Aplicações em Biologia

Strings são sequências imutáveis de caracteres e são fundamentais para representar e manipular dados textuais em programação. Na biologia, strings são amplamente utilizadas para representar sequências biológicas, como DNA, RNA e proteínas. Neste notebook, exploraremos as funcionalidades das strings em Python, com um foco especial em exemplos biológicos.

## 1. O que são Strings?

Uma string é uma sequência de caracteres. Em Python, strings são imutáveis, o que significa que uma vez criadas, elas não podem ser alteradas. Qualquer operação que 'mude' uma string na verdade cria uma nova string. Os caracteres de uma string são indexados, começando do `0`.

## 2. Criando Strings (Sequências Biológicas)

Strings podem ser criadas usando aspas simples (`' '`), aspas duplas (`" "`) ou aspas triplas (`''' '''` ou `""" """`) para strings multi-linha.

```python
# Sequência de DNA simples
sequencia_dna = 'ATGCGTACGTAGCTAGCATGCTAGCT'
print(f"Sequência de DNA: {sequencia_dna}, Tipo: {type(sequencia_dna)}")

# Sequência de RNA
sequencia_rna = "AUGCCUGAGUUUGAA"
print(f"Sequência de RNA: {sequencia_rna}")

# Sequência de proteína (código de 1 letra)
sequencia_proteina = '''MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHGKKVADALTNAVAHVDDMPNALSALSDLHAHKL'''
print(f"Sequência de Proteína (fragmento): {sequencia_proteina[:20]}...")
```

## 3. Acessando Caracteres e Fatiamento (Slicing)

Você pode acessar caracteres individuais em uma string usando seus índices, e subsequências (sub-strings) usando fatiamento (slicing), similar às listas.

```python
dna_completo = 'ATGCGTAGCTAGCTAGCATGCTAGCTAGCTAGCATGCGTAGCTAGCTAGCATGCTAGCT'

# Acessando o primeiro nucleotídeo (índice 0)
print(f"Primeiro nucleotídeo: {dna_completo[0]}")

# Acessando o quinto nucleotídeo (índice 4)
print(f"Quinto nucleotídeo: {dna_completo[4]}")

# Acessando o último nucleotídeo (índice -1)
print(f"Último nucleotídeo: {dna_completo[-1]}")

# Fatiamento: Obter um gene específico (ex: do índice 5 ao 15, exclusivo)
fragmento_gene = dna_completo[5:15]
print(f"Fragmento de gene (5:15): {fragmento_gene}")

# Fatiamento: Obter os primeiros 10 nucleotídeos
primeiros_10 = dna_completo[:10]
print(f"Primeiros 10 nucleotídeos: {primeiros_10}")

# Fatiamento: Obter os últimos 8 nucleotídeos
ultimos_8 = dna_completo[-8:]
print(f"Últimos 8 nucleotídeos: {ultimos_8}")

# Fatiamento com passo (a cada 3 nucleotídeos para frames de leitura, por exemplo)
codons = dna_completo[0::3]
print(f"Códon frame 1: {codons}")
```

### Entendendo o Fatiamento (Slicing) de Strings

O fatiamento de strings em Python é uma ferramenta poderosa para extrair partes (subsequências) de uma string. Ele usa a mesma sintaxe de fatiamento que listas e outras sequências e permite especificar um `início`, um `fim` e um `passo`.

A sintaxe geral é `[início:fim:passo]`:

*   **`início` (opcional)**: O índice onde o fatiamento começa. Se omitido, o padrão é `0` (o início da string).
*   **`fim` (opcional)**: O índice onde o fatiamento *termina*, **mas não inclui** o caractere nesse índice. Se omitido, o padrão é o comprimento da string (até o final).
*   **`passo` (opcional)**: O incremento entre os índices. Se omitido, o padrão é `1`. Um passo negativo inverte a string.

**Pontos Chave:**

*   **Indexação Zero:** Lembre-se que o primeiro caractere está no índice `0`.
*   **Exclusividade do `fim`:** O caractere no índice `fim` *nunca* é incluído no resultado.
*   **Índices Negativos:** Podem ser usados para contar a partir do final da string (`-1` é o último caractere, `-2` o penúltimo, e assim por diante).
*   **Imutabilidade:** O fatiamento *sempre* retorna uma **nova string**. A string original permanece inalterada, reforçando o conceito de que strings são imutáveis em Python.

**Exemplos Adicionais:**

*   `dna[::]` ou `dna[:]`: Retorna uma cópia completa da string.
*   `dna[::2]`: Retorna caracteres em posições pares (0, 2, 4...). Útil para simular uma das fitas de DNA.
*   `dna[::-1]`: Inverte completamente a string. Em biologia, pode ser usado para encontrar a fita complementar reversa de DNA (embora seja preciso substituir os nucleotídeos depois).
*   `dna[10:]`: Do décimo primeiro caractere até o final.
*   `dna[:5]`: Os primeiros cinco caracteres.

```python
nome = "Python"
print(f"Primeiro caractere: {nome[0]}")
print(f"Último caractere: {nome[-1]}")
print(f"Dois primeiros caracteres: {nome[:2]}")
print(f"Dois últimos caracteres: {nome[-2:]}")
print(f"Passando os índeces de 0 até 3: {nome[:3]}")
```

```python
instituto = "Instituto de Biologia Roberto Alcantara Gomes"
print(f"Primeiro caractere: {instituto[0]}")
print(f"Caractere 10: {instituto[9]}")
print(f"De 2 em 2: {instituto[::2]}")
```

## 4. Operações Básicas com Strings (Sequências)

Algumas operações comuns incluem concatenação, repetição e verificação de tamanho.

```python
# Concatenação de sequências (junção)
dna_parte1 = 'ATGC'
dna_parte2 = 'GCTA'
dna_completo = dna_parte1 + dna_parte2
print(f"DNA concatenado: {dna_completo}")

# Repetição de um motivo (ex: sequência repetitiva)
motivo_repetitivo = 'AG' * 5
print(f"Motivo repetitivo: {motivo_repetitivo}")

# Tamanho de uma sequência (comprimento)
sequencia_longa = 'ATGCGTAGCTAGCTAGCATGCTAGCTAGCTAGCATGCGTAGCTAGCTAGCATGCTAGCT'
print(f"Comprimento da sequência: {len(sequencia_longa)} nucleotídeos")

# Verificação de associação (um motivo está na sequência?)
motivo_encontrado = 'TAGC' in sequencia_longa
print(f"'TAGC' está na sequência? {motivo_encontrado}")

motivo_ausente = 'GGGG' in sequencia_longa
print(f"'GGGG' está na sequência? {motivo_ausente}")
```

## 5. Métodos de String com Foco Biológico

Python oferece diversos métodos embutidos para manipulação de strings. Vamos ver alguns úteis para bioinformática.

### Contando Nucleotídeos/Aminoácidos (`.count()`)

Conta o número de ocorrências de uma sub-string (nucleotídeo, códon, etc.) dentro da string principal.

```python
sequencia_exemplo = 'AGCTACGATCGTAGCATCGTAGCATCGA'

# Contar a ocorrência de 'G'
contagem_g = sequencia_exemplo.count('G')
print(f"Número de 'G's: {contagem_g}")

# Contar a ocorrência de um códon 'AGC'
contagem_agc = sequencia_exemplo.count('AGC')
print(f"Número de códons 'AGC': {contagem_agc}")
```

### Encontrando Motivos/Subsequências (`.find()`, `.index()`)

`.find()` retorna o índice da primeira ocorrência da sub-string. Retorna `-1` se não for encontrada. `.index()` faz o mesmo, mas levanta um `ValueError` se não encontrar.

```python
sequencia_promotora = 'ATGCCGTAGATAATGCATGCA'

# Encontrar a posição do TATA box (um motivo comum)
tata_box = 'TATA'
posicao_tata = sequencia_promotora.find(tata_box)
print(f"TATA box encontrado na posição: {posicao_tata}")

# Encontrar um motivo inexistente
motivo_inexistente = 'XXXX'
posicao_inexistente = sequencia_promotora.find(motivo_inexistente)
print(f"Motivo 'XXXX' encontrado na posição: {posicao_inexistente}") # Retorna -1

# Usando .index() (cuidado com erro se não encontrar)
try:
    posicao_gc = sequencia_promotora.index('GCCG')
    print(f"Motivo 'GCCG' encontrado na posição: {posicao_gc}")
except ValueError:
    print(f"Motivo 'GCCG' não encontrado.")
```

### Substituição de Caracteres/Subsequências (`.replace()`)

Útil para simular mutações, converter DNA para RNA (substituindo T por U) ou corrigir erros de sequenciamento.

```python
dna_original = 'ATGCGTACGTAGCTAG'
print(f"DNA original: {dna_original}")

# Convertendo DNA para RNA (T por U)
rna_transcrito = dna_original.replace('T', 'U')
print(f"RNA transcrito: {rna_transcrito}")

# Simular uma mutação pontual (C por G)
dna_mutado = dna_original.replace('C', 'G', 1) # O último argumento é o número máximo de substituições
print(f"DNA mutado (primeiro C por G): {dna_mutado}")
```

### Mudança de Caixa (`.upper()`, `.lower()`, `.capitalize()`)

Útil para padronizar sequências ou cabeçalhos de arquivos FASTA.

```python
sequencia_dna_variada = 'aTgCgTaGcT'

print(f"Original: {sequencia_dna_variada}")
print(f"Maiúsculas: {sequencia_dna_variada.upper()}")
print(f"Minúsculas: {sequencia_dna_variada.lower()}")

cabecalho_fasta = '>gene_de_interesse_humano'
print(f"Cabeçalho capitalizado: {cabecalho_fasta.capitalize()}")
```

### Verificação de Início/Fim (`.startswith()`, `.endswith()`)

Pode ser usada para identificar genes por marcadores de início/fim ou verificar formatos de arquivo.

```python
regiao_genomica = 'ATGCATGCAAGCTAA'

# Verificar se começa com um promotor 'ATGC'
comeca_promotor = regiao_genomica.startswith('ATGC')
print(f"Começa com 'ATGC'? {comeca_promotor}")

# Verificar se termina com um sinal de poliadenilação 'AATAA'
termina_poli_a = regiao_genomica.endswith('AATAA')
print(f"Termina com 'AATAA'? {termina_poli_a}")
```

### Dividir e Juntar Strings (`.split()`, `.join()`)

`.split()` é útil para parsear dados de arquivos (ex: dividir uma linha em campos). `.join()` é útil para construir uma sequência a partir de uma lista de nucleotídeos/aminoácidos.

```python
# Dividindo informações de uma linha de arquivo FASTA
linha_fasta = '>Gene001 | especie=Homo_sapiens | tamanho=1234'
partes = linha_fasta.split(' | ')
print(f"Partes da linha FASTA: {partes}")

# Juntando nucleotídeos de uma lista para formar uma sequência
lista_nucleotideos = ['A', 'T', 'G', 'C', 'G', 'T', 'A']
sequencia_reconstruida = ''.join(lista_nucleotideos)
print(f"Sequência reconstruída: {sequencia_reconstruida}")

# Juntando aminoácidos com um separador (ex: traço)
lista_aminoacidos = ['M', 'V', 'L', 'S', 'P']
proteina_com_separador = '-'.join(lista_aminoacidos)
print(f"Sequência de proteína com separador: {proteina_com_separador}")
```

## 6. F-strings (Strings Formatadas)

F-strings (`f''`) são uma maneira poderosa e legível de incorporar variáveis e expressões em strings, ideal para relatórios e exibição de resultados biológicos.

```python
nome_gene = 'BRCA1'
comprimento_bp = 10000
porcentagem_gc = 45.75

# Usando f-strings para imprimir informações de um gene
print(f"O gene {nome_gene} possui {comprimento_bp} pares de base (pb) e um conteúdo GC de {porcentagem_gc:.2f}%.")

mutacao = 'C->T'
posicao = 12345

print(f"Detectada mutação {mutacao} na posição {posicao} do gene {nome_gene}.")
```

7. ## Operações Aritméticas com Strings

Embora strings não sejam números, podemos realizar operações que se assemelham a operações aritméticas:

*   **Concatenação (`+`)**: Funciona como uma 'adição', juntando duas ou mais strings para formar uma nova.
*   **Repetição (`*`)**: Funciona como uma 'multiplicação', criando uma nova string que é a repetição da original um certo número de vezes.

```python
# padrão
padrão = "=-"
print(f"{padrão * 30}")

# Concatenação (como 'adição' de strings)
parte1 = 'Bio'
parte2 = 'informática'
completo = parte1 + parte2
print(f"Concatenação: {completo}")

# Repetição (como 'multiplicação' de strings)
motivo = 'AG'
repetido = motivo * 3
print(f"Repetição: {repetido}")

# Exemplo com a variável 'nome' existente
nome = "python"
print(f"Repetição de '{nome}': {nome * 2}")

# padrão
padrão = "=-"
print(f"{padrão * 30}")
```

## Imutáveis

As strings em Python são **imutáveis**. Isso significa que, uma vez criada, a sequência de caracteres que compõe a string não pode ser alterada. Qualquer operação que *pareça* modificar uma string (como converter para maiúsculas, fatiar ou substituir caracteres) na verdade **cria e retorna uma nova string**, deixando a original intacta.

**O que isso significa na prática?**

*   Você **não pode** alterar um caractere específico de uma string diretamente por seu índice (ex: `minha_string[0] = 'X'` resultará em um `TypeError`).
*   Métodos como `.replace()`, `.upper()`, `.lower()` não mudam a string original; eles retornam uma *nova string* com as modificações.
*   Para 'atualizar' uma variável de string, você precisa **atribuir uma nova string** a ela.

```python
# criando uma nova variável
linguagem = "python"
linguagem = linguagem.capitalize()
print(linguagem)
```

```python
# Dará erro!
linguagem = "python"
linguagem[0] = "P"
print(linguagem)
```
