## Python para Biologia: Laços de Repetição (Loops)

Na biologia e bioinformática, os laços de repetição, ou *loops*, são ferramentas essenciais. Eles permitem que você automatize a análise de grandes volumes de dados (sequências genéticas, dados de experimentos), simule processos biológicos (crescimento populacional, reações químicas) ou processe informações repetitivas de forma eficiente. Em Python, os dois tipos principais de laços que nos auxiliam nessas tarefas são o `for` e o `while`.

## 1. O Laço `for` na Biologia

O laço `for` é perfeito para percorrer elementos em coleções biológicas, como:
*   **Listas de espécies:** Analisar características de cada espécie.
*   **Sequências genéticas (DNA/RNA):** Processar cada base nitrogenada.
*   **Amostras de dados:** Iterar sobre resultados de experimentos.

Ele é ideal quando você precisa executar uma ação para cada item de uma coleção conhecida.

```python
# Exemplo de 'for' com range() em contextos biológicos

print("Simulando 5 gerações:")
for geracao in range(5): # range(5) gera números de 0 a 4
    print(f"Geração {geracao + 1}") # Ajuste para começar da Geração 1

print("\nProcessando amostras de 1 a 5:")
for amostra_id in range(1, 6): # range(start, stop) gera números de 'start' até 'stop-1'
    print(f"Analisando amostra {amostra_id}")

print("\nContando células em intervalos de 2 micrômetros (0 a 10):")
for tamanho_um in range(0, 11, 2): # range(start, stop, step)
    print(f"Célula com tamanho {tamanho_um} µm")
```

```python
# Exemplo de 'for' com Listas de Espécies e Sequências Genéticas

print("\nAnalisando uma lista de espécies:")
especies_ameacadas = ["Panda Gigante", "Tigre de Bengala", "Gorila da Montanha", "Tartaruga Marinha"]
for especie in especies_ameacadas:
    print(f"A espécie '{especie}' precisa de proteção.")

print("\nIterando sobre uma sequência de DNA (string):")
dna_sequencia = "ATGCGTAGCATCG"
for base in dna_sequencia:
    print(f"Base: {base}")
```

## 2. O Laço `while` para Condições Biológicas

O laço `while` é fundamental quando você precisa que um processo continue *enquanto* uma condição biológica específica for verdadeira. É ideal para cenários onde o número de repetições não é fixo, mas sim determinado por um estado dinâmico:
*   **Simulação de crescimento populacional:** Continuar enquanto a população não excede a capacidade de suporte.
*   **Monitoramento de reações:** Repetir medições enquanto a concentração de um reagente estiver acima de um limite.
*   **Evolução de um traço:** Iterar até que uma característica atinja um certo valor.

```python
# Exemplo de 'while' em simulações biológicas

print("Simulação de decaimento de uma população viral:")
populacao_viral = 1000
dia = 0
while populacao_viral > 100: # Enquanto a população viral for maior que 100
    print(f"Dia {dia}: População viral = {populacao_viral:.0f}") # .0f para remover decimais
    populacao_viral *= 0.8 # Diminui 20% a cada dia
    dia += 1
print(f"Após {dia} dias, a população viral atingiu um nível baixo: {int(populacao_viral)}")

print("\nMonitorando nível de nutrientes até esgotamento (digitando 'esgotou'):")
status_nutriente = ""
while status_nutriente.lower() != "esgotou":
    status_nutriente = input("Nível de nutriente (digite 'esgotou' para parar): ")
    if status_nutriente.lower() != "esgotou":
        print(f"Ainda há nutriente. Medindo novamente...")
print("Nutriente esgotado. Experimento finalizado.")
```

## 3. Controle de Fluxo em Laços Biológicos: `break` e `continue`

Em análises biológicas, podemos precisar controlar o fluxo de um laço de forma mais dinâmica:

*   **`break`:** Interrompe a análise *imediatamente* se uma condição crítica for encontrada (e.g., contaminante detectado, limite de erro excedido).
*   **`continue`:** Pula uma amostra ou dado que não atende aos critérios (e.g., amostra corrompida, indivíduo com mutação irrelevante), passando para a próxima iteração.

```python
# Exemplo de 'break' na análise de sequências

print("Buscando a sequência 'GATC' em um genoma:")
genoma_fragmentado = ["ATGC", "CGTA", "GATC", "TAGC", "CCGG"]
for fragmento in genoma_fragmentado:
    if "GATC" in fragmento:
        print(f"Fragmento 'GATC' encontrado no segmento: {fragmento}! Interrompendo a busca.")
        break # Sai do loop 'for'
    print(f"Analisando fragmento: {fragmento}")

# Exemplo de 'continue' para filtrar dados biológicos

print("\nProcessando apenas amostras com peso acima de 10g:")
pesos_amostras = [5, 12, 8, 15, 6, 20] # em gramas
for peso in pesos_amostras:
    if peso <= 10: # Se o peso for 10g ou menos
        print(f"Amostra com {peso}g ignorada (muito leve).")
        continue # Pula para a próxima iteração
    print(f"Processando amostra com peso válido: {peso}g")
```

## 4. Laços Aninhados para Matrizes e Combinações Biológicas

Um laço dentro de outro laço, ou *laço aninhado*, é extremamente útil em biologia para:
*   **Matrizes de interação:** Representar e analisar interações entre genes, proteínas ou espécies.
*   **Análise de placas de cultura:** Processar dados de cada poço em uma placa de 96 poços.
*   **Gerar combinações:** Criar todas as possíveis combinações de alelos, primers, etc.

```python
# Exemplo de Laços Aninhados em Biologia

print("Interações entre 3 genes e 5 proteínas:")
genes = ["GeneA", "GeneB", "GeneC"]
proteinas = ["Prot1", "Prot2", "Prot3", "Prot4", "Prot5"]

for gene in genes: # Laço externo para cada gene
    for proteina in proteinas: # Laço interno para cada proteína
        print(f"Analisando interação: {gene} <-> {proteina}")
    print("------") # Separador para cada gene

print("\nSimulando uma placa de cultura 3x4:")
num_linhas = 3
num_colunas = 4
for linha in range(num_linhas):
    for coluna in range(num_colunas):
        print(f"Poço ({linha+1},{coluna+1}) ", end="") # +1 para começar do 1
    print() # Quebra de linha após cada linha
```
