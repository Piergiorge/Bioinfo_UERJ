# Python: Laços de Repetição (Loops)

Os laços de repetição, ou *loops*, são uma das ferramentas mais poderosas na programação. Eles permitem que você execute um bloco de código repetidamente, economizando tempo e evitando a duplicação de código. Em Python, os dois tipos principais de laços são o `for` e o `while`.

## 1. O Laço `for`

O laço `for` é usado para iterar sobre uma sequência (como uma lista, tupla, dicionário, conjunto ou string) ou outros objetos iteráveis. Ele é ideal quando você sabe (ou pode determinar) o número de vezes que deseja repetir algo.

```python
# Exemplo de 'for' com range()
print("Contando de 0 a 4:")
for i in range(5): # range(5) gera números de 0 a 4
    print(i)

print("\nContando de 1 a 5:")
for num in range(1, 6): # range(start, stop) gera números de 'start' até 'stop-1'
    print(num)

print("\nContando de 0 a 10, de 2 em 2:")
for j in range(0, 11, 2): # range(start, stop, step)
    print(j)
```

```python
# Exemplo de 'for' com Listas e Strings

print("\nIterando sobre uma lista de frutas:")
frutas = ["maçã", "banana", "cereja", "uva"]
for fruta in frutas:
    print(f"Eu gosto de {fruta}.")

print("\nIterando sobre os caracteres de uma string:")
palavra = "Python"
for letra in palavra:
    print(letra)
```

## 2. O Laço `while`

O laço `while` executa um bloco de código *enquanto* uma condição específica for verdadeira. É ideal quando você não sabe antecipadamente quantas vezes a repetição ocorrerá, mas sabe quando ela deve parar (ou seja, quando a condição se tornar falsa).

```python
# Exemplo de 'while'

print("Contagem regressiva:")
contador = 5
while contador > 0:
    print(contador)
    contador -= 1 # Decrementa o contador em 1
print("Lançamento!")

print("\nPedindo input ao usuário até que ele digite 'sair':")
comando = ""
while comando.lower() != "sair":
    comando = input("Digite um comando (ou 'sair' para finalizar): ")
    if comando.lower() != "sair":
        print(f"Você digitou: {comando}")
print("Programa encerrado.")
```

## 3. Controle de Fluxo em Laços: `break` e `continue`

Duas palavras-chave importantes permitem um controle mais refinado sobre o comportamento dos laços:

*   **`break`:** Interrompe a execução do laço *imediatamente*, saindo dele completamente.
*   **`continue`:** Pula o restante do código na iteração atual e passa para a próxima iteração do laço.

```python
# Exemplo de 'break'

print("Buscando o número 7 na lista (com break):")
numeros = [1, 3, 5, 7, 9, 11]
for num in numeros:
    if num == 7:
        print(f"Número 7 encontrado! Parando o loop.")
        break # Sai do loop 'for'
    print(f"Processando: {num}")

# Exemplo de 'continue'

print("\nImprimindo apenas números pares (com continue):")
for i in range(1, 10):
    if i % 2 != 0: # Se o número for ímpar
        continue # Pula para a próxima iteração sem imprimir
    print(f"Número par: {i}")
```

## 4. Laços Aninhados (Nested Loops)

Um laço pode conter outro laço. Isso é chamado de laço aninhado e é útil para trabalhar com estruturas de dados bidimensionais (como matrizes) ou para gerar combinações.

```python
# Exemplo de Laços Aninhados

print("Tabuada de 1 a 3:")
for i in range(1, 4): # Laço externo para o multiplicador
    for j in range(1, 11): # Laço interno para o multiplicando
        print(f"{i} x {j} = {i * j}")
    print("------") # Separador para cada tabuada

print("\nCriando uma grade simples:")
linhas = 3
colunas = 4
for linha in range(linhas):
    for coluna in range(colunas):
        print(f"({linha},{coluna}) ", end="") # 'end=""' evita quebra de linha
    print() # Quebra de linha após cada linha
```

## Conclusão

Os laços de repetição são pilares da programação, permitindo que você automatize tarefas e processe grandes volumes de dados. O `for` é excelente para iterar sobre coleções, enquanto o `while` é ideal quando a repetição depende de uma condição. O uso de `break` e `continue` oferece flexibilidade no controle do fluxo dos laços, e laços aninhados expandem ainda mais suas capacidades. Dominar esses conceitos é essencial para escrever códigos Python eficientes e poderosos!
