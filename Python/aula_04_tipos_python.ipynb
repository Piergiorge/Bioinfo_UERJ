# Python: Principais Tipos de Dados em Contexto Biológico

A programação em Python, especialmente em bioinformática, depende da manipulação e representação de diversos tipos de informações. Compreender os tipos de dados fundamentais é crucial para armazenar sequências genéticas, resultados de experimentos, características de pacientes e muito mais. Este notebook explorará os principais tipos de dados do Python 3 e como eles se aplicam a cenários biológicos.

## 1. Tipos Numéricos: `int` (Inteiros) e `float` (Ponto Flutuante)

Números são a base de muitas análises biológicas, desde contagens simples até medições precisas. `int` é usado para números inteiros (sem casas decimais), enquanto `float` é para números com casas decimais.

```python
# int: Contagem de cromossomos, número de genes, ciclos de PCR
numero_cromossomos_humanos = 46
contagem_exons_gene = 12
ciclos_pcr = 35

print(f"Cromossomos humanos: {numero_cromossomos_humanos} (Tipo: {type(numero_cromossomos_humanos)})")
print(f"Exons no gene: {contagem_exons_gene} (Tipo: {type(contagem_exons_gene)})")
print(f"Ciclos de PCR: {ciclos_pcr} (Tipo: {type(ciclos_pcr)})")

# float: Concentrações, níveis de expressão, pH, peso molecular
concentracao_dna = 50.75 # ng/uL
nivel_expressao_rna = 1234.56 # RPKM ou FPKM
pH_meio_cultura = 7.2
peso_molecular_proteina = 25.321 # kDa

print(f"\nConcentração de DNA: {concentracao_dna} (Tipo: {type(concentracao_dna)})")
print(f"Nível de expressão de RNA: {nivel_expressao_rna} (Tipo: {type(nivel_expressao_rna)})")
print(f"pH do meio de cultura: {pH_meio_cultura} (Tipo: {type(pH_meio_cultura)})")
print(f"Peso molecular da proteína: {peso_molecular_proteina} (Tipo: {type(peso_molecular_proteina)})")
```

## 2. Tipo Booleano: `bool`

O tipo booleano representa um valor de verdade: `True` (verdadeiro) ou `False` (falso). É fundamental para a lógica condicional, permitindo que seu código tome decisões com base em condições biológicas.

```python
# bool: Gene ativado, presença de mutação, resultado de teste
gene_ativado = True
mutacao_detectada = False
controle_positivo_sucesso = True

print(f"Gene ativado: {gene_ativado} (Tipo: {type(gene_ativado)})")
print(f"Mutação detectada: {mutacao_detectada} (Tipo: {type(mutacao_detectada)})")
print(f"Controle positivo bem-sucedido: {controle_positivo_sucesso} (Tipo: {type(controle_positivo_sucesso)})")

# Uso em uma condição
if gene_ativado and not mutacao_detectada:
    print("\nO gene está funcional e expresso normalmente.")
else:
    print("\nHá uma anomalia na expressão ou sequência do gene.")
```

## 3. Tipos de Sequência: `str` (String), `list` (Lista), `tuple` (Tupla)

Tipos de sequência são coleções ordenadas de itens. Eles são cruciais para armazenar informações como sequências de nucleotídeos, listas de genes ou conjuntos de coordenadas.

### 3.1. String (`str`)

Strings são sequências de caracteres, perfeitas para representar sequências de DNA, RNA, proteínas, nomes de organismos, IDs de amostras e outras informações textuais.

```python
# str: Sequências de DNA/RNA/Proteína, nomes de espécies, IDs de amostras
sequencia_dna = "ATGCGTACGTAGCTAGCATGC"
sequencia_proteina = "MVLSPADKTNVKAAWGKV"
nome_especie = "Escherichia coli"
identificador_amostra = "BIO_EXP_001_REPLICATE_A"

print(f"Sequência de DNA: {sequencia_dna} (Tipo: {type(sequencia_dna)})")
print(f"Comprimento da sequência de DNA: {len(sequencia_dna)}")
print(f"Primeiras 5 bases do DNA: {sequencia_dna[:5]}")

print(f"\nSequência de Proteína: {sequencia_proteina} (Tipo: {type(sequencia_proteina)})")
print(f"Nome da espécie: {nome_especie} (Tipo: {type(nome_especie)})")
```

### 3.2. Lista (`list`)

Listas são coleções ordenadas e **mutáveis** (podem ser alteradas) de itens. Podem conter elementos de diferentes tipos e são ideais para conjuntos de dados que precisam ser modificados, adicionados ou removidos.

```python
# list: Lista de genes, resultados de expressão, amostras em um lote
lista_genes_mutados = ["BRCA1", "TP53", "ATM"]
valores_expressao = [10.5, 23.1, 8.9, 15.2]
linhagens_celulares = ["HEK293", "MCF7", "A549"]

print(f"Genes mutados: {lista_genes_mutados} (Tipo: {type(lista_genes_mutados)})")

# Adicionando um novo gene
lista_genes_mutados.append("KRAS")
print(f"Genes mutados atualizados: {lista_genes_mutados}")

# Removendo um valor
valores_expressao.remove(8.9)
print(f"Valores de expressão atualizados: {valores_expressao}")
```

### 3.3. Tupla (`tuple`)

Tuplas são coleções ordenadas e **imutáveis** (não podem ser alteradas após a criação). São usadas para agrupar dados relacionados que não devem mudar, como coordenadas genômicas (cromossomo, início, fim) ou um conjunto fixo de codons.

```python
# tuple: Coordenadas genômicas, codon de início, informações de alelos fixas
coordenadas_brca1 = ("chr17", 43044295, 43125483) # (cromossomo, start, end)
codon_inicio = ('A', 'T', 'G')

print(f"Coordenadas do BRCA1: {coordenadas_brca1} (Tipo: {type(coordenadas_brca1)})")
print(f"Cromossomo: {coordenadas_brca1[0]}")

print(f"\nCodon de início: {codon_inicio} (Tipo: {type(codon_inicio)})")

# Tentar modificar uma tupla resultaria em um erro:
# coordenadas_brca1[1] = 43044300 # Descomente para testar e ver o erro
```

## 4. Tipo de Mapeamento: `dict` (Dicionário)

Dicionários são coleções **não ordenadas** (em versões anteriores ao Python 3.7, ordenadas a partir de 3.7) e **mutáveis** de pares `chave: valor`. Cada `chave` deve ser única e imutável, e `valor` pode ser de qualquer tipo. Dicionários são excelentes para armazenar informações estruturadas, como anotações de genes, perfis de pacientes ou dados de expressão gênica.

```python
# dict: Anotações de gene, perfil de paciente, genomas de referência

info_gene_tp53 = {
    "nome": "TP53",
    "cromossomo": "chr17",
    "funcao": "Supressor de tumor",
    "mutacoes_conhecidas": ["R175H", "G245S", "R248Q"],
    "nivel_expressao_normal": 150.7
}

perfil_paciente_id = "PAC_042"
perfil_paciente = {
    "idade": 62,
    "sexo": "Feminino",
    "diagnostico": "Câncer de Mama",
    "resposta_ao_tratamento": "Positiva",
    "biomarcadores": {"ER": "Positivo", "PR": "Positivo", "HER2": "Negativo"}
}

print(f"Informações do gene TP53: {info_gene_tp53} (Tipo: {type(info_gene_tp53)})")
print(f"Função do TP53: {info_gene_tp53['funcao']}")

# Adicionando uma nova informação ao gene
info_gene_tp53['isoformas'] = ['alpha', 'beta']
print(f"\nTP53 com isoformas adicionadas: {info_gene_tp53}")

print(f"\nPerfil do Paciente {perfil_paciente_id}: {perfil_paciente} (Tipo: {type(perfil_paciente)})")
print(f"Diagnóstico do paciente: {perfil_paciente['diagnostico']}")
```

## 5. Tipo de Conjunto: `set` (Conjunto)

Conjuntos são coleções **não ordenadas** de **itens únicos**. Eles são úteis para armazenar elementos distintos e para operações matemáticas de conjunto, como união, interseção e diferença (por exemplo, encontrar genes comuns entre dois estudos).

```python
# set: Genes únicos de um experimento, alelos presentes
genes_experimento_A = {"BRCA1", "TP53", "CDKN2A", "KRAS"}
genes_experimento_B = {"TP53", "ATM", "BRCA1", "P53"}

print(f"Genes do Experimento A: {genes_experimento_A} (Tipo: {type(genes_experimento_A)})")

# Adicionando um gene (se já existir, não fará nada)
genes_experimento_A.add("PTEN")
print(f"Genes do Experimento A (após adicionar PTEN): {genes_experimento_A}")

# Interseção: Genes comuns a ambos os experimentos
genes_comuns = genes_experimento_A.intersection(genes_experimento_B)
print(f"\nGenes comuns entre A e B: {genes_comuns}")

# União: Todos os genes únicos dos dois experimentos
todos_os_genes = genes_experimento_A.union(genes_experimento_B)
print(f"Todos os genes únicos em A e B: {todos_os_genes}")

# Diferença: Genes em A que não estão em B
genes_apenas_em_A = genes_experimento_A.difference(genes_experimento_B)
print(f"Genes apenas em A: {genes_apenas_em_A}")
```
