# Python: Laços de Repetição (Loops) em Biologia

Na biologia e bioinformática, muitas tarefas envolvem a repetição de operações. Seja para percorrer uma longa sequência de DNA, processar dados de múltiplas amostras, simular o crescimento de populações bacterianas ou analisar ciclos de reações, os laços de repetição (loops) são ferramentas essenciais para automatizar e otimizar seu código Python.

## 1. Laço `for`

O laço `for` é usado para iterar sobre uma sequência (como uma lista, tupla, string, dicionário ou `range`) ou outros objetos iteráveis. É ideal quando você sabe o número de vezes que deseja repetir uma ação.

```python
# Exemplo 1: Iterando sobre uma Sequência de DNA
sequencia_dna = "ATGCGTACGTAGCTAGCATGC"
print(f"Sequência de DNA: {sequencia_dna}\n")

print("Bases na sequência:")
for base in sequencia_dna:
    print(base)

# Exemplo 2: Contando Bases Específicas
contagem_adenina = 0
for base in sequencia_dna:
    if base == 'A':
        contagem_adenina += 1
print(f"\nNúmero de adeninas (A): {contagem_adenina}")

# Exemplo 3: Processando uma Lista de Genes de Interesse
genes_interesse = ['BRCA1', 'TP53', 'APC', 'KRAS']
print("\nGenes a serem analisados:")
for gene in genes_interesse:
    print(f"Analisando o gene: {gene}")

# Exemplo 4: Iterando sobre um Dicionário (Expressão Gênica)
info_gene_tp53 = {
    'nome': 'TP53',
    'cromossomo': 'chr17',
    'funcao': 'Supressor de tumor',
    'mutacoes_conhecidas': ['R175H', 'G245S', 'R248Q'],
    'nivel_expressao_normal': 150.7
}

print("\nInformações do gene TP53:")
for chave, valor in info_gene_tp53.items():
    print(f"{chave}: {valor}")
```

## 2. Laço `while`

O laço `while` executa um bloco de código repetidamente enquanto uma determinada condição for `True`. É útil quando você não sabe de antemão quantas vezes o laço precisará ser executado, mas sabe quando ele deve parar.

```python
# Exemplo 1: Simulação de Crescimento Bacteriano
populacao_inicial = 50.0 # Bactérias
taxa_crescimento = 1.1 # 10% de aumento por unidade de tempo
limiar_populacional = 500 # Bactérias

tempo = 0
print(f"População inicial: {populacao_inicial:.2f} bactérias\n")

while populacao_inicial < limiar_populacional:
    populacao_inicial *= taxa_crescimento
    tempo += 1
    print(f"Tempo {tempo}: População = {populacao_inicial:.2f}")

print(f"\nCrescimento atingido! População de {populacao_inicial:.2f} alcançada em {tempo} unidades de tempo.")

# Exemplo 2: Estabilização de pH em uma Solução Biológica
ph_solucao = 6.8
pH_neutro = 7.0
contador = 0

print(f"\npH inicial da solução: {ph_solucao}\n")

while abs(ph_solucao - pH_neutro) > 0.01 and contador < 10:
    if ph_solucao < pH_neutro:
        ph_solucao += 0.05 # Adiciona base
    else:
        ph_solucao -= 0.05 # Adiciona ácido
    contador += 1
    print(f"Passo {contador}: pH = {ph_solucao:.2f}")

if abs(ph_solucao - pH_neutro) <= 0.01:
    print(f"\npH estabilizado em {ph_solucao:.2f} após {contador} passos.")
else:
    print(f"\nNão foi possível estabilizar o pH em {contador} passos. pH final: {ph_solucao:.2f}.")
```

## 3. Comandos de Controle de Laço (`break` e `continue`)

Esses comandos permitem um controle mais refinado sobre a execução dos laços:
*   `break`: Interrompe a execução do laço imediatamente e sai dele.
*   `continue`: Pula o restante do código na iteração atual e passa para a próxima iteração do laço.

```python
# Exemplo 1: Usando `break` para Encontrar um Gene Específico
lista_genes_mutados = ['BRCA1', 'TP53', 'ATM', 'KRAS', 'EGFR']
gene_alvo = 'KRAS'

print("\nBuscando gene alvo na lista de mutações...")
for gene in lista_genes_mutados:
    print(f"Verificando {gene}...")
    if gene == gene_alvo:
        print(f"Gene '{gene_alvo}' encontrado! Interrompendo a busca.")
        break
else: # O 'else' do for é executado se o loop terminar sem um 'break'
    print(f"Gene '{gene_alvo}' não encontrado na lista.")

# Exemplo 2: Usando `continue` para Ignorar Amostras Contaminadas
identificadores_amostras = ['Amostra_001', 'Amostra_CONTAMINADA_002', 'Amostra_003', 'Amostra_CONTAMINADA_004', 'Amostra_005']

print("\nProcessando amostras (ignorando contaminadas):")
for amostra in identificadores_amostras:
    if 'CONTAMINADA' in amostra:
        print(f"Amostra {amostra} está contaminada. Pulando para a próxima.")
        continue
    print(f"Processando dados da {amostra}...")
print("Processamento de amostras concluído.")
```

## 4. Laços Aninhados (Nested Loops)

Laços aninhados são laços dentro de outros laços. Eles são úteis quando você precisa iterar sobre combinações de elementos de diferentes coleções ou para trabalhar com estruturas de dados multidimensionais, como matrizes de expressão gênica ou interações proteína-proteína.

```python
# Exemplo: Comparando Linhagens Celulares com Condições de Cultivo
linhagens_celulares = ['HEK293', 'MCF7', 'A549']
condicoes_cultivo = ['DME-HighGlucose', 'RPMI-1640', 'DMEM-F12']

print("\nCombinando Linhagens Celulares com Condições de Cultivo:")
for linhagem in linhagens_celulares:
    for condicao in condicoes_cultivo:
        print(f"Experimentando {linhagem} na condição {condicao}")

# Exemplo 2: Gerando todas as combinações de codons de início (A, T, G)
codon_bases = ['A', 'T', 'G']
print("\nPossíveis códons de início (A, T, G):")
for b1 in codon_bases:
    for b2 in codon_bases:
        for b3 in codon_bases:
            print(f"{b1}{b2}{b3}")
```
