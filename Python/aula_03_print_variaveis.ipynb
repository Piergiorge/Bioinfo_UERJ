# Python: A Função `print()` e Variáveis

Neste notebook, vamos explorar dois conceitos fundamentais em Python: a função `print()`, usada para exibir informações, e as variáveis, que servem para armazenar dados. Abordaremos esses tópicos com exemplos práticos inspirados no campo da biologia, tornando o aprendizado mais relevante e aplicado.

## 1. A Função `print()`: Exibindo Dados Biológicos

A função `print()` é usada para enviar saída para o console. É uma ferramenta essencial para depurar código, mostrar resultados de análises ou simplesmente apresentar informações ao usuário. Vamos ver como ela pode ser utilizada para exibir dados biológicos.

### 1.1. Uso Básico da `print()`

Podemos imprimir strings literais, números, ou o conteúdo de variáveis.

```python
# Imprimindo uma mensagem simples sobre biologia
print("Bem-vindo ao mundo da Bioinformática com Python!")

# Imprimindo um número (ex: número de genes)
print(10000)

# Imprimindo o resultado de uma expressão
print(2 * 5000) # Número total de bases, por exemplo
```

### 1.2. Imprimindo Múltiplos Itens

A função `print()` pode receber múltiplos argumentos, que serão impressos separados por um espaço por padrão.

```python
# Dados biológicos para impressão
nome_proteina = "Hemoglobina"
numero_aminoacidos = 141

# Imprimindo múltiplos itens
print("Nome da proteína:", nome_proteina, "Número de aminoácidos:", numero_aminoacidos)

# Você pode alterar o separador (sep) e o final (end) da linha
print("DNA", "RNA", "Proteína", sep=' -> ')
print("Início da análise... ", end='')
print("Concluído!")
```

### 1.3. F-strings (Strings Formatadas) para Relatórios Biológicos

As f-strings são a forma mais moderna e recomendada para incorporar variáveis e expressões em strings. Elas são muito úteis para criar relatórios legíveis sobre dados biológicos.

```python
# Dados de uma amostra de sequenciamento
identificador_amostra = "AMOSTRA_001"
especie = "Homo sapiens"
sequencia_lida = "ATGCGTAGCTAGCTAGCATGC"
tamanho_sequencia = len(sequencia_lida)
potencial_mutacao = "C>T"

# Usando f-strings para apresentar os dados
print(f"Análise da amostra: {identificador_amostra}")
print(f"Espécie detectada: {especie}")
print(f"Sequência inicial (primeiros 10bp): {sequencia_lida[:10]}...")
print(f"Tamanho total da sequência: {tamanho_sequencia} pares de base (pb)")
print(f"Mutação detectada em potencial: {potencial_mutacao} na posição 5.")

# Com cálculos embutidos
percentual_gc = (sequencia_lida.count('G') + sequencia_lida.count('C')) / tamanho_sequencia * 100
print(f"Conteúdo GC: {percentual_gc:.2f}%") # Formatação para 2 casas decimais
```

## 2. Variáveis: Armazenando Informações

Variáveis são espaços nomeados na memória do computador que armazenam dados. Elas permitem que você manipule e reutilize informações em seu código. Em biologia, podemos usar variáveis para armazenar sequências, nomes de genes, contagens de expressão, etc.

### 2.1. Criação e Atribuição de Variáveis

Para criar uma variável, você atribui um valor a um nome.

```python
# Atribuindo uma sequência de DNA a uma variável
sequencia_dna = "ATGCGTACGTAGCTAGCATGCTAGCT"

# Atribuindo o nome de um gene
nome_gene = "BRCA1"

# Atribuindo um valor numérico (ex: nível de expressão)
nivel_expressao = 150.75

# Atribuindo um valor booleano (ex: gene ativado)
gene_ativado = True

print(f"Sequência DNA: {sequencia_dna}")
print(f"Gene: {nome_gene}")
print(f"Nível de Expressão: {nivel_expressao}")
print(f"Gene Ativado: {gene_ativado}")
```

### 2.2. Regras de Nomenclatura para Variáveis

- Devem começar com uma letra (a-z, A-Z) ou um underscore (`_`).
- Não podem começar com um número.
- Podem conter letras, números e underscores.
- São case-sensitive (`nome_gene` é diferente de `Nome_gene`).
- Palavras reservadas do Python não podem ser usadas como nomes de variáveis (ex: `if`, `for`, `print`).

```python
# Exemplos de nomes de variáveis válidos em bioinformática

_primeira_sequencia = "AAAAATTTTT"
contagem_exons = 5
GenomaHumano = "GRCh38"

print(f"Variável válida 1: {_primeira_sequencia}")
print(f"Variável válida 2: {contagem_exons}")
print(f"Variável válida 3: {GenomaHumano}")

# Exemplos de nomes de variáveis inválidos (descomente para ver o erro)
# 1gene = "ERBB2" # Começa com número
# for = "loop" # Palavra reservada
```

### 2.3. Verificando o Tipo de Variáveis

A função `type()` é útil para verificar o tipo de dado que uma variável está armazenando.

```python
sequencia_proteica = "MVLSPADKTNVKAAWGKV"
ponto_mutacao = 123
probabilidade_doenca = 0.85
esta_infectado = False

print(f"Tipo de 'sequencia_proteica': {type(sequencia_proteica)}")
print(f"Tipo de 'ponto_mutacao': {type(ponto_mutacao)}")
print(f"Tipo de 'probabilidade_doenca': {type(probabilidade_doenca)}")
print(f"Tipo de 'esta_infectado': {type(esta_infectado)}")
```

### 2.4. Reatribuindo Valores a Variáveis

Você pode alterar o valor de uma variável a qualquer momento. Ela pode até mesmo mudar de tipo de dado.

```python
status_experimento = "Iniciado"
print(f"Status inicial: {status_experimento}")

status_experimento = "Em andamento"
print(f"Status atualizado: {status_experimento}")

# A variável pode mudar de tipo
status_experimento = True
print(f"Status final (booleano): {status_experimento}, Tipo: {type(status_experimento)}")
```
