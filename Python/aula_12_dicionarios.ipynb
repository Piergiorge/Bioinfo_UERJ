# Python: Dicionários

---

Na biologia e bioinformática, é comum lidarmos com dados que possuem uma relação chave-valor, onde cada informação pode ser identificada por um nome ou um identificador único. Pense nas propriedades de um gene, como seu nome, cromossomo, função e nível de expressão; ou nas características de uma amostra, como seu ID, espécie e data de coleta. Para organizar e acessar esses dados de forma eficiente, os **dicionários** em Python são a estrutura de dados perfeita.

## 1. O que são Dicionários?

Um dicionário (`dict`) em Python é uma coleção **não ordenada** e **mutável** de itens. Cada item em um dicionário é um par chave-valor (`key-value pair`).

*   **Chaves (Keys)**: Devem ser únicas e imutáveis (strings, números, tuplas).
*   **Valores (Values)**: Podem ser de qualquer tipo de dado e podem se repetir.
*   São definidos por chaves `{}` e os pares chave-valor são separados por vírgulas, com a chave e o valor separados por dois pontos `:`.

## 2. Criando Dicionários

Podemos criar dicionários vazios ou pré-populados com pares chave-valor.

```python
# Exemplo 1: Informações de um Gene
info_gene_tp53 = {
    'nome': 'TP53',
    'cromossomo': 'chr17',
    'funcao': 'Supressor de tumor',
    'mutacoes_conhecidas': ['R175H', 'G245S', 'R248Q'],
    'nivel_expressao_normal': 150.7
}
print(f"Informações do Gene TP53: {info_gene_tp53}")

# Exemplo 2: Perfil de uma Amostra Biológica
perfil_paciente = {
    'id': 'PAC_042',
    'idade': 62,
    'sexo': 'Feminino',
    'diagnostico': 'Câncer de Mama',
    'resposta_ao_tratamento': 'Positiva'
}
print(f"\nPerfil do Paciente: {perfil_paciente}")

# Exemplo 3: Dicionário vazio
dados_analise_sequenciamento = {}
print(f"\nDicionário vazio para dados de sequenciamento: {dados_analise_sequenciamento}, {type(dados_analise_sequenciamento)}")
```

## 3. Acessando Elementos

Os valores em um dicionário são acessados usando suas chaves. Há duas maneiras principais: usando colchetes `[]` ou o método `get()`.

```python
info_gene_tp53 = {
    'nome': 'TP53',
    'cromossomo': 'chr17',
    'funcao': 'Supressor de tumor',
    'mutacoes_conhecidas': ['R175H', 'G245S', 'R248Q'],
    'nivel_expressao_normal': 150.7
}

# Acessando um valor usando colchetes
nome_gene_exemplo = info_gene_tp53['nome']
print(f"Nome do gene: {nome_gene_exemplo}")

# Acessando um valor usando get()
func_gene = info_gene_tp53.get('funcao')
print(f"Função do gene: {func_gene}")

# Tentando acessar uma chave inexistente:
# Com colchetes: Gerará um KeyError
# print(info_gene_tp53['localizacao'])

# Com get(): Retorna None (ou um valor padrão se especificado), sem erro
localizacao = info_gene_tp53.get('localizacao', 'Não especificado')
print(f"Localização (usando get com valor padrão): {localizacao}")
```

## 4. Modificando Dicionários

Dicionários são mutáveis, permitindo adicionar novos pares chave-valor, atualizar valores existentes ou remover itens.

```python
perfil_paciente = {
    'id': 'PAC_042',
    'idade': 62,
    'sexo': 'Feminino',
    'diagnostico': 'Câncer de Mama'
}
print(f"Perfil inicial: {perfil_paciente}")

# Adicionar um novo par chave-valor
perfil_paciente['resposta_ao_tratamento'] = 'Positiva'
print(f"\nApós adicionar resposta ao tratamento: {perfil_paciente}")

# Atualizar um valor existente
perfil_paciente['idade'] = 63
print(f"Após atualizar a idade: {perfil_paciente}")

# Adicionar um dicionário aninhado (biomarcadores)
perfil_paciente['biomarcadores'] = {'ER': 'Positivo', 'PR': 'Positivo', 'HER2': 'Negativo'}
print(f"Após adicionar biomarcadores: {perfil_paciente}")

# Remover um par chave-valor (usando del)
del perfil_paciente['sexo']
print(f"Após remover 'sexo' com del: {perfil_paciente}")

# Remover um par chave-valor e retornar seu valor (usando pop)
valor_removido = perfil_paciente.pop('diagnostico')
print(f"Após remover 'diagnostico' com pop (valor: {valor_removido}): {perfil_paciente}")

# Remover o último item inserido (em Python 3.7+) e retornar o par (popitem)
item_removido = perfil_paciente.popitem()
print(f"Após remover o último item com popitem (item: {item_removido}): {perfil_paciente}")

# Limpar todos os elementos do dicionário
perfil_paciente.clear()
print(f"Após limpar o dicionário: {perfil_paciente}")
```

## 5. Métodos Úteis de Dicionários

Dicionários possuem diversos métodos que facilitam a manipulação de seus conteúdos.

```python
info_gene_tp53 = {
    'nome': 'TP53',
    'cromossomo': 'chr17',
    'funcao': 'Supressor de tumor',
    'mutacoes_conhecidas': ['R175H', 'G245S', 'R248Q'],
    'nivel_expressao_normal': 150.7,
}

# keys(): Retorna uma visualização de todas as chaves
chaves = info_gene_tp53.keys()
print(f"Chaves do dicionário: {chaves}")

# values(): Retorna uma visualização de todos os valores
valores = info_gene_tp53.values()
print(f"Valores do dicionário: {valores}")

# items(): Retorna uma visualização dos pares chave-valor
itens = info_gene_tp53.items()
print(f"Itens do dicionário: {itens}")

# update(): Combina dicionários ou atualiza valores
novas_infos = {'localizacao': 'Núcleo', 'nivel_expressao_normal': 160.0}
info_gene_tp53.update(novas_infos)
print(f"\nApós atualizar com novas_infos: {info_gene_tp53}")

# setdefault(): Insere uma chave com um valor padrão se a chave não existir
# Se 'cofator' não existir, adiciona-o com o valor 'N/A'
cofactor = info_gene_tp53.setdefault('cofator', 'N/A')
print(f"Cofator (adicionado se não existir): {cofactor}")
print(f"Dicionário após setdefault: {info_gene_tp53}")

# Se 'nome' já existe, setdefault retorna seu valor e não o altera
nome_existente = info_gene_tp53.setdefault('nome', 'GENE_DEFAULT')
print(f"Nome existente (setdefault não altera): {nome_existente}")
print(f"Dicionário após setdefault para chave existente: {info_gene_tp53}")
```

## 6. Iterando sobre Dicionários

É muito comum precisar iterar sobre os dicionários para processar seus dados. Podemos iterar sobre as chaves, os valores ou os pares chave-valor.

```python
info_gene_tp53 = {
    'nome': 'TP53',
    'cromossomo': 'chr17',
    'funcao': 'Supressor de tumor',
    'mutacoes_conhecidas': ['R175H', 'G245S', 'R248Q'],
    'nivel_expressao_normal': 150.7
}

print("Iterando sobre as chaves:")
for chave in info_gene_tp53:
    print(f"Chave: {chave}")

print("\nIterando sobre os valores:")
for valor in info_gene_tp53.values():
    print(f"Valor: {valor}")

print("\nIterando sobre os pares chave-valor (itens):")
for chave, valor in info_gene_tp53.items():
    print(f"{chave}: {valor}")

# Exemplo: Filtrando informações com base em uma condição
print("\nInformações relevantes para o gene:")
limiar_ativacao = 100.0
for chave, valor in info_gene_tp53.items():
    if chave == 'nivel_expressao_normal' and valor > limiar_ativacao:
        print(f"O gene {info_gene_tp53['nome']} tem um nível de expressão ativado: {valor}")
    elif chave != 'mutacoes_conhecidas': # Exibir outras informações importantes, exceto a lista de mutações
        print(f"{chave}: {valor}")
```
