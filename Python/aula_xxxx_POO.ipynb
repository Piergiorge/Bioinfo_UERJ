## Programação Orientada a Objetos (POO) em Biologia

A Programação Orientada a Objetos (POO) é um paradigma de programação que organiza o software em torno de **objetos**, em vez de funções e lógica. Um objeto pode ser definido como uma instância de uma **classe**, que é um "molde" ou "projeto" que descreve o comportamento e as propriedades que seus objetos terão.

## Conceitos Chave da POO:

1.  **Classes:** Moldes para criar objetos. Elas definem um conjunto de atributos (características) e métodos (comportamentos) que os objetos criados a partir delas terão.
2.  **Objetos:** Instâncias de classes. Quando você cria um objeto a partir de uma classe, ele possui os atributos e métodos definidos por essa classe.
3.  **Atributos:** Variáveis que pertencem a uma classe ou objeto, representando suas características ou estado.
4.  **Métodos:** Funções que pertencem a uma classe ou objeto, representando suas ações ou comportamentos.
5.  **Herança:** Permite que uma classe (subclasse) herde atributos e métodos de outra classe (superclasse), promovendo a reutilização de código e a criação de hierarquias.
6.  **Encapsulamento:** Oculta os detalhes internos de um objeto, expondo apenas o que é necessário para interagir com ele. Isso protege os dados e simplifica o uso do objeto.
7.  **Polimorfismo:** Permite que objetos de diferentes classes sejam tratados de forma uniforme se compartilharem uma interface comum. Por exemplo, diferentes objetos podem responder a um mesmo método de maneiras diferentes.
8.  **Abstração:** Foca na representação de características essenciais de um objeto, ocultando detalhes de implementação complexos. Isso simplifica o design e o uso de classes e objetos.

## Por que POO em Biologia?

A biologia é inerentemente complexa, com entidades como genes, proteínas, células, amostras e experimentos, cada um com suas próprias características e comportamentos. A POO é ideal para modelar essa complexidade, permitindo:

*   **Organização e Clareza:** Representar conceitos biológicos como classes (`Gene`, `Proteina`, `Amostra`), tornando o código mais legível e estruturado.
*   **Reusabilidade:** Criar classes genéricas e estendê-las para casos mais específicos (por exemplo, uma `SequenciaDNA` pode ser estendida para `Gene`).
*   **Manutenção:** Facilita a modificação e expansão de funcionalidades, pois as responsabilidades estão bem definidas dentro dos objetos.
*   **Modelagem de Dados:** Capturar as propriedades e relações de entidades biológicas de forma intuitiva.

```python
print("### Exemplo 1: Modelando uma Sequência de DNA ###")

class DNASequence:
    def __init__(self, sequence, identifier):
        # Atributos da classe, prefixados com '_' para indicar que são internos
        self._sequence = sequence.upper()  # Armazena a sequência em maiúsculas
        self._id = identifier

    def get_length(self):
        # Método para obter o comprimento da sequência
        return len(self._sequence)

    def transcribe(self):
        # Método para transcrever a sequência de DNA para RNA
        # Substitui 'T' por 'U'
        return self._sequence.replace('T', 'U')

    def complement_reverse(self):
        # Método para obter o complemento reverso da sequência de DNA
        complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
        complement_seq = "".join([complement_map[base] for base in self._sequence])
        return complement_seq[::-1] # Inverte a sequência

    def get_gc_content(self):
        # Método para calcular o conteúdo GC
        gc_count = self._sequence.count('G') + self._sequence.count('C')
        return (gc_count / self.get_length()) * 100 if self.get_length() > 0 else 0

    def __str__(self):
        # Método especial para representação em string do objeto
        return f"DNASequence(ID: {self._id}, Length: {self.get_length()}, Sequence: {self._sequence[:10]}...{self._sequence[-10:]})"

# Criando objetos (instâncias) da classe DNASequence
dna_seq1 = DNASequence("ATGCGTACGTAGCTAGCATGCATGCATGCATGCATGC", "Seq_001")
dna_seq2 = DNASequence("GATCAGATCTAGATCGATCGATCGA", "Seq_002")

print(f"\nObjeto 1: {dna_seq1}")
print(f"Comprimento da {dna_seq1._id}: {dna_seq1.get_length()} bases")
print(f"RNA Transcrito: {dna_seq1.transcribe()[:10]}...")
print(f"Complemento Reverso: {dna_seq1.complement_reverse()[:10]}...")
print(f"Conteúdo GC: {dna_seq1.get_gc_content():.2f}%")

print(f"\nObjeto 2: {dna_seq2}")
print(f"Comprimento da {dna_seq2._id}: {dna_seq2.get_length()} bases")
print(f"RNA Transcrito: {dna_seq2.transcribe()[:10]}...")
print(f"Complemento Reverso: {dna_seq2.complement_reverse()[:10]}...")
print(f"Conteúdo GC: {dna_seq2.get_gc_content():.2f}%")
```

### Exemplo 2: Herança - Criando uma Classe `Gene` a partir de `DNASequence`

A herança nos permite criar classes mais específicas (`Gene`) que reutilizam a funcionalidade de classes mais genéricas (`DNASequence`). Um `Gene` é uma `DNASequence` com propriedades adicionais, como nome e função.

```python
print("### Exemplo 2: Herança - Classe Gene ###")

class Gene(DNASequence):
    def __init__(self, sequence, identifier, name, function, organism="Homo sapiens"):
        # Chama o construtor da classe pai (DNASequence)
        super().__init__(sequence, identifier)
        self._name = name
        self._function = function
        self._organism = organism

    def get_gene_info(self):
        # Método específico da classe Gene
        return (
            f"Gene ID: {self._id}\n"
            f"Nome: {self._name}\n"
            f"Organismo: {self._organism}\n"
            f"Função: {self._function}\n"
            f"Comprimento da sequência: {self.get_length()} bases"
        )

    def __str__(self):
        # Sobrescreve o método __str__ para Gene
        return f"Gene(ID: {self._id}, Nome: {self._name}, Organismo: {self._organism}, Comprimento: {self.get_length()})"

# Criando um objeto da classe Gene
tp53_gene = Gene(
    "ATGGAGCCGCAGTCAGATCCTAGAAACTTCAGTGTAGATTTAGAATTAGTAGAACTACA",
    "ENSG00000141510",
    "TP53",
    "Supressor de tumor, regulador do ciclo celular",
    "Homo sapiens"
)

print(f"\nObjeto Gene: {tp53_gene}")
print("\nInformações do Gene TP53:")
print(tp53_gene.get_gene_info())
print(f"Conteúdo GC do Gene TP53: {tp53_gene.get_gc_content():.2f}%") # Método herdado
```

### Exemplo 3: Modelando uma Sequência de Proteína

Assim como DNA, podemos modelar sequências de proteínas como objetos, adicionando métodos específicos para suas propriedades, como peso molecular.

```python
print("### Exemplo 3: Classe ProteinSequence ###")

class ProteinSequence:
    def __init__(self, sequence, identifier):
        self._sequence = sequence.upper()
        self._id = identifier

    def get_length(self):
        return len(self._sequence)

    def calculate_molecular_weight(self):
        # Exemplo simplificado de cálculo de peso molecular
        # Valores médios de peso molecular para cada aminoácido (em Da)
        aa_weights = {
            'A': 89.09, 'R': 174.20, 'N': 132.12, 'D': 133.10, 'C': 121.16,
            'E': 147.13, 'Q': 146.15, 'G': 75.07, 'H': 155.16, 'I': 131.18,
            'L': 131.18, 'K': 146.19, 'M': 149.21, 'F': 165.19, 'P': 115.13,
            'S': 105.09, 'T': 119.12, 'W': 204.23, 'Y': 181.19, 'V': 117.15
        }
        total_weight = 0
        for aa in self._sequence:
            total_weight += aa_weights.get(aa, 0) # Adiciona 0 se o AA não for encontrado
        return total_weight

    def __str__(self):
        return f"ProteinSequence(ID: {self._id}, Length: {self.get_length()}, Sequence: {self._sequence[:10]}...{self._sequence[-10:]})"

# Criando um objeto ProteinSequence
hemoglobin_alpha = ProteinSequence(
    "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHGKKVADALTNAVAHVDDMPNALSALSDLHAHKLRVDPVNFKLLSHCLLVTLAAHLPAEFTPAVHASLDKFLASVSTVLTSKYR",
    "Hemoglobin_alpha_chain"
)

print(f"\nObjeto Proteína: {hemoglobin_alpha}")
print(f"Comprimento da cadeia alfa da Hemoglobina: {hemoglobin_alpha.get_length()} aminoácidos")
print(f"Peso Molecular (aprox.): {hemoglobin_alpha.calculate_molecular_weight():.2f} Da")
```
