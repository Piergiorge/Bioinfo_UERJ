## 2. Operadores Lógicos em Biologia (Tabelas Verdade)

Os operadores lógicos `and`, `or` e `not` são usados para combinar e manipular condições biológicas. As tabelas verdade nos ajudam a entender o resultado de diferentes combinações.

## 1. Tipo Booleano (`bool`): Verdadeiro ou Falso

O tipo `bool` representa valores de verdade e falsidade. Na biologia, podemos pensar em características ou estados que são claramente 'verdadeiros' ou 'falsos'. Por exemplo, um organismo 'é vertebrado' (True) ou 'não é vertebrado' (False). Em Python, existem duas constantes booleanas:
- `True`: representa verdadeiro.
- `False`: representa falso.

Ambas as constantes são escritas com a primeira letra maiúscula.

```python
# Exemplos de valores booleanos em biologia

organismo_vertebrado = True
tem_clorofila = False

print(f"O organismo é vertebrado: {organismo_vertebrado}, Tipo: {type(organismo_vertebrado)}")
print(f"O organismo tem clorofila: {tem_clorofila}, Tipo: {type(tem_clorofila)}")

# Expressões que resultam em booleanos para dados biológicos
quantidade_oxigenio = 21 # em %
temperatura_ambiente = 25 # em Celsius

print(f"Oxigênio > 20%: {quantidade_oxigenio > 20}") # Nível de oxigênio ideal para muitas espécies
print(f"Temperatura == 25: {temperatura_ambiente == 25}") # Temperatura ótima para algumas enzimas
print(f"'Bactéria' == 'vírus': {'Bactéria' == 'vírus'}") # Comparação de tipos de microrganismos
```

## 2. Operadores Lógicos (Tabelas Verdade)

Os operadores lógicos `and`, `or` e `not` são usados para combinar e manipular valores booleanos. As tabelas verdade mostram o resultado de todas as combinações possíveis.

### 2.1. Operador `and` (E Lógico): Ambas as Condições Verdadeiras

O operador `and` retorna `True` se *ambas* as condições forem verdadeiras. Caso contrário, retorna `False`.

**Tabela Verdade para `and`:**

| A     | B     | A and B |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |

```python
# Exemplos com 'and' em biologia

# Condições para ser um mamífero aquático
tem_glandulas_mamarias = True
vive_na_agua = True

print(f"Tem glândulas mamárias E vive na água: {tem_glandulas_mamarias and vive_na_agua}")

# Condições para uma planta realizar fotossíntese eficientemente
tem_luz_suficiente = True
tem_agua_disponivel = False

print(f"Tem luz suficiente E água disponível: {tem_luz_suficiente and tem_agua_disponivel}")

# Exemplo prático: Classificação de um organismo
eh_animal = True
tem_celulas_eucarioticas = True
tem_parede_celular = False

if eh_animal and tem_celulas_eucarioticas and not tem_parede_celular:
    print("Provavelmente é um organismo do Reino Animalia.")
else:
    print("Não se encaixa perfeitamente na descrição de um animal típico.")
```

### 2.2. Operador `or` (OU Lógico): Pelo Menos Uma Condição Verdadeira

O operador `or` retorna `True` se *pelo menos uma* das condições for verdadeira. Retorna `False` apenas se ambas as condições forem falsas.

**Tabela Verdade para `or`:**

| A     | B     | A or B  |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | True    |
| False | True  | True    |
| False | False | False   |

```python
# Exemplos com 'or' em biologia

# Condições para ser considerado um organismo voador (ex: pássaro ou inseto)
tem_asas = True
e_leve_o_suficiente = True

print(f"Tem asas OU é leve o suficiente: {tem_asas or e_leve_o_suficiente}")

# Condições para uma espécie ser classificada como invasora (se um dos critérios for atingido)
domina_nicho = False
causa_dano_ecologico = True

if domina_nicho or causa_dano_ecologico:
    print("Esta espécie pode ser considerada invasora.")
else:
    print("Esta espécie não parece ser invasora com base nesses critérios.")
```

### 2.3. Operador `not` (NÃO Lógico): Invertendo o Estado

O operador `not` inverte o valor booleano da condição. Se for `True`, torna-se `False`, e vice-versa. Por exemplo, `not tem_doenca` significa 'não tem a doença'.

```python
# Exemplos com 'not' em biologia

tem_mutacao_gene_x = False

print(f"Não tem mutação no gene X: {not tem_mutacao_gene_x}")

dados_genoma_analisados = False

if not dados_genoma_analisados:
    print("Os dados do genoma ainda precisam ser analisados.")
else:
    print("Análise do genoma concluída.")
```

### 2.4 Operadores Lógicos (`and`, `or`, `not`)
Os operadores lógicos permitem combinar e manipular valores booleanos para criar condições mais complexas:
*   `and`: Retorna `True` se **ambas** as condições forem `True`.
*   `or`: Retorna `True` se **pelo menos uma** das condições for `True`.
*   `not`: Inverte o valor booleano ( `True` se torna `False`, e vice-versa).

```python
# Exemplos de Operadores Lógicos em Biologia

# Cenário 1: Diagnóstico de Doença (AND)
sintomas_presentes = True
exame_laboratorial_positivo = True
doenca_detectada = sintomas_presentes and exame_laboratorial_positivo
print(f"\nDoença detectada (sintomas E exame positivo): {doenca_detectada}")

# Cenário 2: Indicação de Tratamento (OR)
mutacao_gene_X = True
alto_risco_hereditario = False
indicacao_terapia_alvo = mutacao_gene_X or alto_risco_hereditario
print(f"Indicação de terapia alvo (mutação OU alto risco): {indicacao_terapia_alvo}")

# Cenário 3: Ausência de Contaminação (NOT)
cultura_bacteriana_contaminada = False
cultura_pura = not cultura_bacteriana_contaminada
print(f"Cultura pura (NÃO contaminada): {cultura_pura}")

# Combinando Operadores
ph_solucao = 7.4
e_solucao_neutra = ph_solucao == 7.0
e_gene_A_ativado = True

condicao_pesquisa = (not e_solucao_neutra) and e_gene_A_ativado
print(f"Condição para pesquisa (solução NÃO neutra E gene ativado): {condicao_pesquisa}")
```

## 3. Estruturas Condicionais: `if`, `elif`, `else`

As estruturas condicionais permitem que você execute diferentes blocos de código com base na avaliação de condições. Isso é essencial para criar programas que modelam processos biológicos, classificam organismos ou analisam dados com base em critérios.

### 3.1. `if` Statement: Reagindo a uma Única Condição

O bloco `if` executa um conjunto de instruções **se** a condição booleana for `True`. Por exemplo, 'se a temperatura estiver acima de um certo limite, um processo é ativado'.

```python
# Exemplo de 'if' em biologia

temperatura_ambiente = 38 # em Celsius

if temperatura_ambiente > 37:
    print(f"A temperatura está acima do ideal para a maioria das enzimas humanas!")
```

### 3.2. `if`...`else` Statement: Duas Possibilidades

O bloco `if`...`else` executa um conjunto de instruções **se** a condição for `True`, e um bloco alternativo de instruções **se** a condição for `False`. Pense em 'se o gene X está presente, faça A; senão, faça B'.

```python
# Exemplo de 'if'...'else' em biologia

pH_solucao = 6.5

if pH_solucao < 7.0:
    print("A solução é ácida.")
else:
    print("A solução é neutra ou básica.")
```

### 3.3. `if`...`elif`...`else` Statement: Múltiplas Condições

Use `elif` (abreviação de 'else if') para verificar múltiplas condições. O Python avalia as condições em ordem e executa o bloco de código da primeira condição que for `True`. Se nenhuma condição `if` ou `elif` for `True`, o bloco `else` (se existir) será executado. Isso é útil para classificação de espécies, análise de diferentes estados de uma doença, etc.

```python
# Exemplo de 'if'...'elif'...'else' em biologia: Classificação de Glicemia

nivel_glicose = 95 # em mg/dL

if nivel_glicose < 70:
    print("Estado: Hipoglicemia")
elif nivel_glicose >= 70 and nivel_glicose <= 99:
    print("Estado: Glicemia Normal")
elif nivel_glicose >= 100 and nivel_glicose <= 125:
    print("Estado: Pré-diabetes")
else:
    print("Estado: Diabetes")

# Outro exemplo: Classificação de tipo de célula
tipo_celula = "eucariótica animal"

if tipo_celula == "eucariótica vegetal":
    print("Célula com parede celular e cloroplastos.")
elif tipo_celula == "eucariótica animal":
    print("Célula sem parede celular, com centríolos.")
elif tipo_celula == "procariótica":
    print("Célula sem núcleo definido e organelas membranosas.")
else:
    print("Tipo de célula não reconhecido.")
```

```python
# Cenário Biológico: Análise de Risco Genético
score_risco_doenca = 0.75
limiar_alto_risco = 0.7
limiar_medio_risco = 0.4

print(f"\nScore de Risco Genético: {score_risco_doenca}")

if score_risco_doenca >= limiar_alto_risco:
    print("Paciente com alto risco para a doença. Recomenda-se acompanhamento especializado.")
elif score_risco_doenca >= limiar_medio_risco:
    print("Paciente com risco médio para a doença. Monitoramento regular é aconselhado.")
else:
    print("Paciente com baixo risco para a doença. Seguir recomendações gerais de saúde.")


# Cenário Biológico: Reação de PCR
temperatura_reacao = 55 # Exemplo de temperatura em °C
temperatura_ideal_annealing = 58
ciclos_pcr = 30

print(f"\nTemperatura da Reação: {temperatura_reacao}°C, Ciclos: {ciclos_pcr}")

if temperatura_reacao < temperatura_ideal_annealing - 5:
    print("Atenção: Temperatura de anelamento muito baixa. Pode haver ligação inespecífica.")
elif temperatura_reacao > temperatura_ideal_annealing + 5:
    print("Atenção: Temperatura de anelamento muito alta. Pode reduzir a eficiência.")
elif ciclos_pcr < 25:
    print("Número de ciclos de PCR baixo. Repetir com mais ciclos para maior amplificação.")
elif ciclos_pcr > 35:
    print("Número de ciclos de PCR alto. Risco de artefatos de amplificação.")
else:
    print("Condições da reação de PCR ideais. Proseguir com análise.")


# Cenário Biológico: Verificação de Expressão Gênica e Mutação para Tratamento
nivel_expressao_gene = 180.2 # Unidades arbitrárias
gene_com_mutacao = True

print(f"\nNível de Expressão do Gene: {nivel_expressao_gene}, Gene com Mutação: {gene_com_mutacao}")

if nivel_expressao_gene > 150 and gene_com_mutacao:
    print("Recomendar tratamento X devido a alta expressão E mutação presente.")
elif nivel_expressao_gene > 150 and not gene_com_mutacao:
    print("Considerar tratamento Y devido a alta expressão, mas sem mutação.")
elif nivel_expressao_gene <= 150 and gene_com_mutacao:
    print("Analisar impacto da mutação, pois expressão é normal.")
else:
    print("Não há indicação clara para tratamentos específicos baseados nestes critérios.")
```

### 3.4. Aninhamento de Condicionais: Lógicas  Mais Complexas

Você pode aninhar estruturas condicionais dentro de outras para criar lógicas mais complexas, simulando a hierarquia de decisões ou características.

```python
# Exemplo de condicionais aninhadas em biologia: Classificação de um animal

e_mamifero = True
produz_leite = True
tem_pelos = False

if e_mamifero:
    print("É um mamífero.")
    if produz_leite:
        print("Este mamífero produz leite para seus filhotes.")
        if tem_pelos:
            print("E também possui pelos.")
        else:
            print("Mas não possui pelos visíveis (ex: golfinho).")
    else:
        print("Parece ser um mamífero atípico que não produz leite (erro na lógica ou característica rara?).")
else:
    print("Não é um mamífero.")
```
