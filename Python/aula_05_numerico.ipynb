# Python: Tipos Numéricos em Contexto Biológico (Inteiros, Flutuantes e Complexos)

Na biologia e bioinformática, números são onipresentes. Seja para contar genes, medir concentrações, calcular pesos moleculares ou modelar processos biológicos, a manipulação numérica é fundamental. O Python oferece três tipos numéricos principais para lidar com essas diversas representações: `int` (inteiros), `float` (números de ponto flutuante) e `complex` (números complexos).

## 1. Inteiros (`int`)

Representam números inteiros, sem casas decimais. São ideais para contagens discretas, como número de genes, ciclos de reação, número de amostras ou contagens de bases nitrogenadas.

```python
# Exemplos de Inteiros em Biologia
numero_genes = 25000
numero_cromossomos = 23 # Pares em humanos
ciclos_pcr = 30
contagem_amostras = 100

print(f"Número total de genes: {numero_genes} (Tipo: {type(numero_genes)})")
print(f"Pares de cromossomos: {numero_cromossomos} (Tipo: {type(numero_cromossomos)})")
print(f"Ciclos de PCR executados: {ciclos_pcr} (Tipo: {type(ciclos_pcr)})")
print(f"Amostras coletadas: {contagem_amostras} (Tipo: {type(contagem_amostras)})")

# Operações Matemáticas com Inteiros
amostras_por_grupo = contagem_amostras // 5 # Divisão inteira
mutacoes_detectadas = 5
novas_mutacoes = 2
total_mutacoes = mutacoes_detectadas + novas_mutacoes

print(f"\nAmostras por grupo (divisão inteira): {amostras_por_grupo}")
print(f"Total de mutações registradas: {total_mutacoes}")
print(f"Diferença de ciclos de PCR: {ciclos_pcr - 10}")
print(f"Multiplicação para cópias de DNA: {2 ** 5} (2 elevado à 5ª potência)")
```

## 2. Ponto Flutuante (`float`)

Representam números com casas decimais. São indispensáveis para medições que envolvem precisão, como concentrações de substâncias, pH, níveis de expressão gênica, pesos moleculares, temperaturas ou probabilidades.

```python
# Exemplos de Floats em Biologia
concentracao_proteina = 0.875 # mg/mL
pH_solucao = 7.42
temperatura_reacao = 37.5 # °C
probabilidade_mutacao = 0.0012
peso_molecular = 64.123 # kDa

print(f"Concentração de proteína: {concentracao_proteina} (Tipo: {type(concentracao_proteina)})")
print(f"pH da solução tampão: {pH_solucao} (Tipo: {type(pH_solucao)})")
print(f"Temperatura da reação enzimática: {temperatura_reacao} (Tipo: {type(temperatura_reacao)})")
print(f"Probabilidade de mutação: {probabilidade_mutacao} (Tipo: {type(probabilidade_mutacao)})")
print(f"Peso molecular da hemoglobina: {peso_molecular} (Tipo: {type(peso_molecular)})")

# Operações Matemáticas com Floats
volume_amostra_ml = 0.5
concentracao_final = concentracao_proteina * volume_amostra_ml

novo_ph = pH_solucao + 0.1 # Ajuste de pH

percentual_gc = (45.0 / 100.0) * 100 # Exemplo de cálculo percentual

print(f"\nQuantidade de proteína em 0.5 mL: {concentracao_final:.3f} mg")
print(f"Novo pH ajustado: {novo_ph:.2f}")
print(f"Percentual GC calculado: {percentual_gc:.2f}%")
print(f"Diferença de temperatura: {temperatura_reacao - 25.0} °C")
```

## 3. Números Complexos (`complex`)

Números complexos são compostos por uma parte real e uma parte imaginária (denotada por `j` no Python, e não `i`). Embora menos comuns na bioinformática clássica, podem surgir em áreas mais avançadas, como processamento de sinais biológicos (EEG, ECG) ou modelagem de fenômenos com componentes de fase.

```python
# Exemplos de Complexos em Biologia (Uso mais nichado)
# Representação de impedância bioelétrica ou análise de fase de ondas
impedancia_tecido = 10 + 5j # Parte real: resistência, Parte imaginária: reatância
onda_cerebral_fase = 0.8 + 1.2j # Exemplo hipotético de componente de fase

print(f"Impedância do tecido: {impedancia_tecido} (Tipo: {type(impedancia_tecido)})")
print(f"Componente de fase da onda cerebral: {onda_cerebral_fase} (Tipo: {type(onda_cerebral_fase)})")

# Operações Matemáticas com Complexos
soma_impedancias = impedancia_tecido + (2 + 3j)
produto_ondas = onda_cerebral_fase * (0.5 - 0.1j)

print(f"\nSoma de impedâncias: {soma_impedancias}")
print(f"Produto de componentes de ondas: {produto_ondas}")
print(f"Parte real da impedância: {impedancia_tecido.real}")
print(f"Parte imaginária da impedância: {impedancia_tecido.imag}")
```

Para ilustrar como os valores apresentados na imagem seriam atualizados em Python, considere os seguintes exemplos. Cada parâmetro tem um valor original que é modificado por uma operação específica para gerar um novo valor.

```python
# 1. Atualização da Concentração de Proteína
concentracao_proteina_original = 0.875 # mg/mL
atualizacao_proteina = 0.1 # mg/mL
concentracao_proteina_nova = concentracao_proteina_original + atualizacao_proteina
print(f"Concentração de Proteína: {concentracao_proteina_original} -> +{atualizacao_proteina} -> {concentracao_proteina_nova:.3f} mg/mL")

# 2. Atualização do pH da Solução
ph_solucao_original = 7.42
atualizacao_ph = -0.05
ph_solucao_novo = ph_solucao_original + atualizacao_ph
print(f"pH da Solução: {ph_solucao_original} -> {atualizacao_ph} -> {ph_solucao_novo:.2f}")

# 3. Atualização da Temperatura da Reação
temperatura_reacao_original = 37.5 # °C
multiplicador_temperatura = 1.1
temperatura_reacao_nova = temperatura_reacao_original * multiplicador_temperatura
print(f"Temperatura da Reação: {temperatura_reacao_original} -> x{multiplicador_temperatura} -> {temperatura_reacao_nova:.2f} °C")
```

Estes exemplos demonstram como operações aritméticas simples em Python podem ser usadas para simular as 'Atualizações' e derivar os 'Novos Valores' mostrados na tabela da imagem, refletindo a manipulação de dados em cenários biológicos.

Além de atualizarmos os valores em novas variáveis, é possível fazer isso diretamente em uma única variável.

```python
x = 20
x = x + 20 # ou x += 20
print(x)

mult = 20 * 5
mult = mult * 2 # ou mult *= 2
print(mult)
```
