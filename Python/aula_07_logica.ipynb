## 1. Variáveis e Tipos de Dados em Biologia

Variáveis são 'contêineres' para armazenar dados. Em biologia, podemos armazenar sequências de DNA, níveis de expressão gênica, informações de pacientes, etc. Python possui vários tipos de dados para diferentes tipos de informação.

```python
# Variáveis para armazenar dados biológicos

# String (sequência de DNA, nome de gene)
sequencia_dna_exemplo = "ATGCTAGCTAGCTACGATGCAT"
nome_gene_exemplo = "TP53"

# Integer (número de exons, contagem de amostras)
numero_exons = 12
contagem_amostras = 25

# Float (nível de expressão, concentração de proteína)
nivel_expressao = 125.75
concentracao_proteina = 0.893 # mg/mL

# Boolean (gene mutado, presença de doença)
gene_mutado = True
paciente_infectado = False

print(f"Sequência de DNA: {sequencia_dna_exemplo}")
print(f"Gene: {nome_gene_exemplo}")
print(f"Número de exons: {numero_exons}")
print(f"Nível de expressão: {nivel_expressao}")
print(f"Gene mutado: {gene_mutado}")
print(f"Tipo de 'sequencia_dna_exemplo': {type(sequencia_dna_exemplo)}")
print(f"Tipo de 'numero_exons': {type(numero_exons)}")
print(f"Tipo de 'nivel_expressao': {type(nivel_expressao)}")
print(f"Tipo de 'gene_mutado': {type(gene_mutado)}")
```

## 2. Operadores em Contexto Biológico

Operadores são símbolos que realizam operações em valores e variáveis. Temos operadores aritméticos, de comparação e lógicos, todos úteis para manipular e comparar dados biológicos.

### 2.1. Operadores Aritméticos

```python
# Cálculo de proporções e quantidades

# Contagem de nucleotídeos
contagem_adenina = 150
contagem_timina = 120
contagem_guanina = 180
contagem_citosina = 130

comprimento_total = contagem_adenina + contagem_timina + contagem_guanina + contagem_citosina
print(f"Comprimento total da sequência: {comprimento_total} pb")

percentual_gc = ((contagem_guanina + contagem_citosina) / comprimento_total) * 100
print(f"Percentual GC: {percentual_gc:.2f}%")

# Divisão de amostras em grupos
numero_amostras_totais = 30
numero_grupos = 3
amostras_por_grupo = numero_amostras_totais // numero_grupos # Divisão inteira
print(f"Amostras por grupo: {amostras_por_grupo}")
```

### 2.2. Operadores de Comparação

Usados para comparar dois valores e retornar `True` ou `False`. Essencial para tomada de decisões com base em dados biológicos (ex: se um nível de expressão está acima de um limiar).

```python
# Comparando níveis de expressão ou contagens

nivel_expressao_gene_A = 150.5
limiar_ativacao = 100.0

e_gene_A_ativado = nivel_expressao_gene_A > limiar_ativacao
print(f"O gene A está ativado? {e_gene_A_ativado}")

pH_solucao = 7.4
pH_neutro = 7.0

e_solucao_neutra = pH_solucao == pH_neutro
print(f"A solução é neutra? {e_solucao_neutra}")

ponto_mutacao_esperado = 123
ponto_mutacao_encontrado = 123

e_mutacao_na_posicao_esperada = ponto_mutacao_esperado == ponto_mutacao_encontrado
print(f"A mutação foi encontrada na posição esperada? {e_mutacao_na_posicao_esperada}")
```

### 2.3. Operadores Lógicos

Combinam expressões booleanas (`True` ou `False`) para formar condições mais complexas. (`and`, `or`, `not`).

```python
# Combinando condições para análise de dados biológicos

gene_ativado = True
proteina_expressa = True

# Um processo ocorre se o gene e a proteína estiverem ativos
processo_ocorre = gene_ativado and proteina_expressa
print(f"O processo biológico ocorre? {processo_ocorre}")

# Uma doença é detectada se o paciente estiver infectado OU tiver mutação
paciente_infectado = False
gene_com_mutacao = True

doenca_detectada = paciente_infectado or gene_com_mutacao
print(f"Doença detectada? {doenca_detectada}")

# O experimento não pode continuar se o status não for 'pronto'
status_experimento_pronto = False

pode_continuar = not status_experimento_pronto
print(f"O experimento pode continuar? {pode_continuar}")
```

## 3. Estruturas Condicionais (`if`, `elif`, `else`)

Permitem que seu programa tome decisões com base em condições. Em biologia, isso pode ser usado para classificar amostras, aplicar diferentes análises dependendo de um valor, etc.

```python
# Exemplo: Classificando o nível de expressão de um gene

nivel_expressao_gene = 180.2

if nivel_expressao_gene > 200:
    print("Gene altamente expresso.")
elif nivel_expressao_gene > 100:
    print("Gene moderadamente expresso.")
else:
    print("Gene com baixa expressão.")

# Exemplo: Tomada de decisão em um diagnóstico

score_risco_doenca = 0.75 # de 0 a 1
limiar_alto_risco = 0.7
limiar_medio_risco = 0.4

if score_risco_doenca >= limiar_alto_risco:
    print("Paciente com alto risco de doença. Recomendar acompanhamento intensivo.")
elif score_risco_doenca >= limiar_medio_risco:
    print("Paciente com risco médio de doença. Recomendar monitoramento regular.")
else:
    print("Paciente com baixo risco de doença. Recomendar estilo de vida saudável.")
```

## 4. Estruturas de Repetição (`for`, `while`)

Loops permitem executar um bloco de código repetidamente. Essencial para processar grandes conjuntos de dados biológicos, como iteração sobre nucleotídeos de uma sequência, amostras em um experimento, ou genes em um genoma.

### 4.1. Loop `for`

```python
# Iterando sobre uma sequência de DNA
sequencia_mrna = "AUGCGUACGUAGCAUGC"

print("Contagem de bases no mRNA:")
contagem_bases = {'A': 0, 'U': 0, 'G': 0, 'C': 0}
for base in sequencia_mrna:
    contagem_bases[base] += 1
print(contagem_bases)

# Iterando sobre uma lista de genes de interesse
genes_interesse = ["BRCA1", "TP53", "APC", "KRAS"]

print("\nVerificando genes:")
for gene in genes_interesse:
    if gene == "BRCA1":
        print(f"O gene {gene} é um gene supressor de tumor importante.")
    else:
        print(f"Analisando gene: {gene}")
```

### 4.2. Loop `while`

```python
# Exemplo: Simulação de crescimento populacional de bactérias até atingir um limiar

populacao_inicial = 100
taxa_crescimento = 1.1 # 10% de aumento por ciclo
limiar_populacional = 500
ciclos = 0

print("Simulação de crescimento bacteriano:")
while populacao_inicial < limiar_populacional:
    populacao_inicial *= taxa_crescimento
    ciclos += 1
    print(f"Ciclo {ciclos}: População = {int(populacao_inicial)}")

print(f"A população atingiu {int(populacao_inicial)} bactérias em {ciclos} ciclos.")
```

## 5. Funções: Organizando o Código Biológico

Funções são blocos de código reutilizáveis que realizam uma tarefa específica. Elas são fundamentais para organizar seu código, torná-lo mais legível e evitar repetição, o que é crucial em projetos de bioinformática complexos.

```python
# Função para calcular o conteúdo GC de uma sequência de DNA
def calcular_conteudo_gc(sequencia_dna):
    """
    Calcula o percentual de G e C em uma sequência de DNA.
    """
    sequencia_upper = sequencia_dna.upper()
    contagem_g = sequencia_upper.count('G')
    contagem_c = sequencia_upper.count('C')
    comprimento = len(sequencia_upper)

    if comprimento == 0:
        return 0.0

    percentual = (contagem_g + contagem_c) / comprimento * 100
    return percentual

# Função para converter DNA em RNA (transcrição)
def transcrever_dna_para_rna(sequencia_dna):
    """
    Converte uma sequência de DNA para sua respectiva sequência de RNA.
    """
    return sequencia_dna.upper().replace('T', 'U')

# Exemplos de uso das funções
sequencia_1 = "ATGCGTACGTAGCTAGCATGC"
sequencia_2 = "AAATTTCCCGGG"

print(f"Conteúdo GC da sequência 1: {calcular_conteudo_gc(sequencia_1):.2f}%")
print(f"Conteúdo GC da sequência 2: {calcular_conteudo_gc(sequencia_2):.2f}%")

rna_sequencia_1 = transcrever_dna_para_rna(sequencia_1)
print(f"Sequência DNA: {sequencia_1} -> RNA: {rna_sequencia_1}")
```
