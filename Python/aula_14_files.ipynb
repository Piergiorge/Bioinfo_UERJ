# Python: Leitura e Escrita de Arquivos em Biologia

Em bioinformática e biologia computacional, a capacidade de ler e escrever arquivos é fundamental. Seja para processar sequências de DNA de um arquivo FASTA, analisar dados de expressão gênica em um CSV, ou salvar resultados de uma análise, o Python oferece ferramentas robustas para interagir com o sistema de arquivos. Nesta seção, exploraremos como realizar operações básicas de I/O (Input/Output) de arquivos, com exemplos práticos e uma observação importante sobre o ambiente Google Colab.

## 1. Abrindo e Fechando Arquivos

Para interagir com um arquivo, primeiro você precisa abri-lo. A função `open()` é utilizada para isso. É crucial sempre fechar um arquivo após o uso para liberar recursos e garantir que todas as escritas sejam salvas. A melhor prática é usar a instrução `with open()`, que garante que o arquivo será fechado automaticamente, mesmo que ocorra um erro.

````python
# Criando um arquivo de exemplo para demonstração
# No Google Colab, este arquivo será criado no ambiente temporário da VM.
# Localmente, ele seria criado na mesma pasta do seu script Python.

with open('sequencia_dna.txt', 'w') as f:
    f.write('>GeneA_sequence\n')
    f.write('ATGCGTACGTAGCTAGCATGC\n')
    f.write('>GeneB_sequence\n')
    f.write('AAATTTCCCGGG')
print("Arquivo 'sequencia_dna.txt' criado com sucesso.")

# O 'with open()' garante que o arquivo 'f' seja automaticamente fechado.
````

### Modos de Abertura:
*   `'r'` (read): Leitura (padrão). O arquivo deve existir.
*   `'w'` (write): Escrita. Cria um novo arquivo ou sobrescreve um existente. **Cuidado!**
*   `'a'` (append): Anexar. Adiciona conteúdo ao final do arquivo; cria o arquivo se ele não existir.
*   `'x'` (exclusive creation): Criação exclusiva. Cria um novo arquivo, mas gera um erro se o arquivo já existir.
*   `'b'` (binary): Modo binário (usado com `r`, `w`, `a` para arquivos não textuais, ex: `rb`, `wb`).
*   `'t'` (text): Modo texto (padrão, pode ser omitido). Usado com `r`, `w`, `a` para arquivos textuais.

## 2. Lendo Arquivos

Existem várias maneiras de ler o conteúdo de um arquivo. Escolha o método mais adequado ao tamanho e formato do seu arquivo.

### **Importante sobre o Google Colab:**

Ao trabalhar no Google Colab, se você tem arquivos que residem no seu computador local ou no Google Drive, você precisa **fazer o upload** deles para o ambiente de execução do Colab antes de poder lê-los com `open()`. Para fazer isso:

2.  **Montar Google Drive:** Para persistência ou para arquivos maiores, você pode montar seu Google Drive. Use o seguinte código para isso:

    ```python

    # Agora você pode acessar seus arquivos em '/content/drive/My Drive/'
    ```

Em um ambiente de desenvolvimento local (como VS Code, PyCharm), você simplesmente coloca o arquivo na mesma pasta do seu script Python ou especifica o caminho completo para ele.

````python
# Lendo o arquivo inteiro
print("\n--- Conteúdo completo do arquivo ---")
with open('sequencia_dna.txt', 'r') as f:
    conteudo_completo = f.read()
    print(conteudo_completo)

# Lendo linha por linha com um loop (muito comum para FASTA, CSV)
print("\n--- Lendo linha por linha ---")
sequencias = {}
identificador_atual = ''
with open('sequencia_dna.txt', 'r') as f:
    for linha in f:
        linha = linha.strip() # Remove quebras de linha e espaços em branco
        if linha.startswith('>'):
            identificador_atual = linha[1:] # Remove o '>' inicial
            sequencias[identificador_atual] = ''
        else:
            sequencias[identificador_atual] += linha

for id_seq, seq in sequencias.items():
    print(f"ID: {id_seq}, Sequência: {seq}")


# Lendo todas as linhas em uma lista
print("\n--- Lendo todas as linhas em uma lista ---")
with open('sequencia_dna.txt', 'r') as f:
    linhas_lista = f.readlines()
    print(linhas_lista)
````

## 3. Escrevendo Arquivos

Escrever arquivos é igualmente importante para salvar resultados de análises ou gerar novos arquivos de dados.

````python
# Escrevendo uma nova sequência em um novo arquivo
print("\n--- Escrevendo nova sequência ---")
nova_sequencia_id = '>GeneC_mutant'
nova_sequencia = 'GGCCTTAA'

with open('nova_sequencia.fasta', 'w') as f_out:
    f_out.write(f"{nova_sequencia_id}\n")
    f_out.write(f"{nova_sequencia}\n")
print("Arquivo 'nova_sequencia.fasta' criado.")

# Anexando dados a um arquivo existente
print("\n--- Anexando dados ao arquivo existente ---")
with open('sequencia_dna.txt', 'a') as f_append:
    f_append.write('\n>GeneD_variant\n')
    f_append.write('TACGTACG')
print("Dados anexados a 'sequencia_dna.txt'.")

# Verificando o conteúdo atualizado de 'sequencia_dna.txt'
print("\n--- Conteúdo atualizado de 'sequencia_dna.txt' ---")
with open('sequencia_dna.txt', 'r') as f:
    print(f.read())
````

## 4. Exemplo com Dados Biológicos: Processando um CSV de Expressão Gênica

Vamos simular a leitura de um arquivo CSV contendo dados de expressão gênica e realizar uma análise simples.

````python
# Criar um arquivo CSV de exemplo
csv_data = [
    'Gene,Sample1,Sample2,Sample3\n',
    'BRCA1,150.5,160.2,145.8\n',
    'TP53,90.1,88.5,92.3\n',
    'MYC,250.3,270.1,265.0\n',
    'KRAS,50.7,52.1,49.9\n'
]

with open('expressao_genica.csv', 'w') as f_csv:
    f_csv.writelines(csv_data)
print("Arquivo 'expressao_genica.csv' criado.")

# Lendo e processando o CSV
dados_expressao = {}
with open('expressao_genica.csv', 'r') as f_csv:
    header = f_csv.readline().strip().split(',')
    samples = header[1:] # Nomes das amostras

    for linha in f_csv:
        partes = linha.strip().split(',')
        gene_nome = partes[0]
        valores_str = partes[1:]
        # Converter valores de string para float
        valores_expressao_float = [float(val) for val in valores_str]
        dados_expressao[gene_nome] = dict(zip(samples, valores_expressao_float))

print("\nDados de expressão gênica processados:")
for gene, expressoes in dados_expressao.items():
    print(f"Gene: {gene}, Expressões: {expressoes}")

# Encontrando o gene com maior expressão média
maior_media = 0
gene_maior_media = ''
for gene, expressoes in dados_expressao.items():
    media_expressao = sum(expressoes.values()) / len(expressoes)
    if media_expressao > maior_media:
        maior_media = media_expressao
        gene_maior_media = gene

print(f"\nGene com a maior expressão média: {gene_maior_media} (Média: {maior_media:.2f})")
````
