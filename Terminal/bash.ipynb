## Principais Comandos Linux: Guia para Colab e Terminal Local

Este notebook tem como objetivo apresentar os comandos Linux mais fundamentais e demonstrar como utilizá-los tanto no ambiente do Google Colab quanto em um terminal local.

### O que são Comandos Linux?

Comandos Linux são instruções de texto que você digita em um terminal (ou shell) para interagir com o sistema operacional. Eles permitem realizar uma vasta gama de tarefas, desde navegar por diretórios e manipular arquivos até gerenciar processos e configurar o sistema. A linha de comando é uma ferramenta poderosa para automação e controle preciso.

### Como Executar Comandos Linux no Google Colab

No Google Colab, você pode executar comandos Linux diretamente de uma célula de código Python prefixando o comando com um ponto de exclamação (`!`). O Colab roda em um ambiente Linux (Ubuntu), então a maioria dos comandos funcionará como esperado.

````python
# Exemplo: listar o conteúdo do diretório atual no Colab
!ls -l
````

### Como Executar Comandos Linux em um Terminal Local

Em um terminal local (seja no Linux, macOS ou WSL no Windows), você simplesmente digita o comando e pressiona `Enter`. Não há necessidade do `!`.

**Exemplo (Terminal Local):**
```bash
ls -l
```

### Comandos Essenciais

Vamos explorar alguns dos comandos Linux mais úteis.

#### 1. `pwd` (Print Working Directory)

Mostra o caminho completo do diretório atual onde você está trabalhando.

````python
# No Colab
!pwd
````

#### 2. `ls` (List)

Lista o conteúdo de um diretório. Várias opções podem ser usadas:
- `ls`: Lista o conteúdo do diretório atual.
- `ls -l`: Lista o conteúdo em formato longo (detalhes como permissões, proprietário, tamanho, data).
- `ls -a`: Lista todos os arquivos, incluindo os ocultos (que começam com `.`).
- `ls -lh`: Formato longo com tamanhos de arquivo legíveis por humanos (KB, MB, GB).

````python
# Listar conteúdo do diretório atual
!ls

# Listar com detalhes
!ls -l

# Listar tudo, incluindo ocultos
!ls -a
````

#### 3. `cd` (Change Directory)

Permite navegar entre diretórios.
- `cd nome_do_diretorio`: Entra em um subdiretório.
- `cd ..`: Volta um diretório (para o diretório pai).
- `cd ~`: Vai para o diretório home do usuário (atalho).
- `cd /`: Vai para o diretório raiz.

````python
# Criar um diretório de exemplo para navegação
!mkdir meu_diretorio_teste

# Entrar no diretório (o !cd não muda o diretório persistente do kernel no Colab,
# mas sim do subshell que executa o comando. Para mudar de diretório de fato em células Python,
# usamos o comando mágico %cd, que é um recurso do IPython/Jupyter).
%cd meu_diretorio_teste

# Verificar o diretório atual
!pwd

# Voltar um diretório
%cd ..

# Verificar novamente
!pwd

# Limpar o diretório de teste
!rmdir meu_diretorio_teste
````

#### 4. `mkdir` (Make Directory)

Cria um novo diretório (pasta).

````python
# Criar um novo diretório
!mkdir nova_pasta

# Verificar se foi criado
!ls

# Remover a pasta para limpeza
!rmdir nova_pasta
````

#### 5. `touch`

Cria um novo arquivo vazio ou atualiza a data e hora de acesso/modificação de um arquivo existente.

````python
# Criar um arquivo vazio
!touch meu_arquivo.txt

# Verificar se foi criado
!ls -l

# Remover o arquivo para limpeza
!rm meu_arquivo.txt
````

#### 6. `cp` (Copy)

Copia arquivos e diretórios.
- `cp origem destino`: Copia um arquivo `origem` para um `destino`.
- `cp -r origem_diretorio destino_diretorio`: Copia um diretório recursivamente.

````python
# Criar um arquivo para copiar
!echo "Este é o conteúdo do arquivo original." > arquivo_original.txt

# Copiar o arquivo
!cp arquivo_original.txt copia_do_arquivo.txt

# Verificar os arquivos
!ls

# Remover os arquivos
!rm arquivo_original.txt copia_do_arquivo.txt
````

#### 7. `mv` (Move)

Move ou renomeia arquivos e diretórios.
- `mv origem destino`: Move `origem` para `destino`.
- `mv nome_antigo.txt nome_novo.txt`: Renomeia um arquivo.

````python
# Criar um arquivo para mover/renomear
!touch arquivo_antigo.txt

# Renomear o arquivo
!mv arquivo_antigo.txt arquivo_novo.txt

# Verificar
!ls

# Remover o arquivo
!rm arquivo_novo.txt
````

#### 8. `rm` (Remove)

Remove arquivos e diretórios.
- `rm arquivo.txt`: Remove um arquivo.
- `rm -r diretorio`: Remove um diretório e seu conteúdo recursivamente (cuidado com este comando!).
- `rm -rf diretorio`: Força a remoção de um diretório e seu conteúdo recursivamente sem pedir confirmação (**EXTREMO CUIDADO!**).


````python
# Criar um arquivo para remover
!touch arquivo_para_remover.txt

# Remover o arquivo
!rm arquivo_para_remover.txt

# Verificar
!ls

# Criar um diretório com um arquivo dentro para testar rm -r
!mkdir pasta_para_remover
!touch pasta_para_remover/arquivo_na_pasta.txt

# Remover o diretório e seu conteúdo
!rm -r pasta_para_remover

# Verificar
!ls
````

#### 9. `cat` (Concatenate)

Exibe o conteúdo de um arquivo. Também pode ser usado para concatenar arquivos ou criar um arquivo digitando seu conteúdo diretamente.

````python
# Criar um arquivo com conteúdo
!echo "Linha 1 do texto." > meu_texto.txt
!echo "Linha 2 do texto." >> meu_texto.txt # O >> adiciona ao arquivo, > sobrescreve

# Exibir o conteúdo
!cat meu_texto.txt

# Remover o arquivo
!rm meu_texto.txt
````

#### 10. `grep` (Global Regular Expression Print)

Procura por padrões de texto dentro de arquivos. É extremamente útil para filtrar logs ou encontrar informações específicas.

````python
# Criar um arquivo com várias linhas
!echo "apple" > frutas.txt
!echo "banana" >> frutas.txt
!echo "orange" >> frutas.txt
!echo "grape" >> frutas.txt
!echo "pineapple" >> frutas.txt

# Procurar por 'apple' no arquivo
!grep "apple" frutas.txt

# Procurar por 'na' (case-sensitive)
!grep "na" frutas.txt

# Procurar por 'na' (case-insensitive) com -i
!grep -i "na" frutas.txt

# Remover o arquivo
!rm frutas.txt
````

#### 11. `head`

Exibe as primeiras linhas de um arquivo (por padrão, as 10 primeiras).

````python
# Criar um arquivo com mais de 10 linhas
!seq 1 15 > numeros.txt # Cria um arquivo com números de 1 a 15, cada um em uma linha

# Exibir as 10 primeiras linhas
!head numeros.txt

# Exibir as 5 primeiras linhas com -n 5
!head -n 5 numeros.txt

# Remover o arquivo
!rm numeros.txt
````

#### 12. `tail`

Exibe as últimas linhas de um arquivo (por padrão, as 10 últimas).

````python
# Usando o mesmo arquivo de antes, se ainda existir, ou recriando
!seq 1 15 > numeros.txt

# Exibir as 10 últimas linhas
!tail numeros.txt

# Exibir as 3 últimas linhas com -n 3
!tail -n 3 numeros.txt

# Remover o arquivo
!rm numeros.txt
````

#### 13. `man` (Manual)

Exibe o manual de ajuda para um comando. É a melhor forma de aprender todas as opções e usos de qualquer comando Linux.

````python
# Exibir o manual para o comando 'ls'
# No Colab, o 'man' pode não exibir interativamente. Geralmente, ele imprime o manual completo.
# Em um terminal local, 'man' abre um visualizador interativo.

# Para ver um trecho no Colab, podemos redirecionar a saída para 'head'
!man ls | head -n 20
````

#### 17. `tr` (Translate or Delete Characters)

Substitui ou deleta caracteres de um texto. Muito útil para formatação e limpeza de strings.

````python
# Converter letras minúsculas para maiúsculas
!echo "hello world" | tr 'a-z' 'A-Z'

# Substituir espaços por hífens
!echo "hello world" | tr ' ' '-'

# Deletar todos os 'o's
!echo "hello world" | tr -d 'o'
````

#### 18. `paste`

Combina linhas de múltiplos arquivos, lado a lado, ou múltiplos campos de um único arquivo.

````python
# Criar arquivos de exemplo
!echo -e "apple\nbanana\norange" > frutas1.txt
!echo -e "red\nyellow\norange" > cores1.txt

# Combinar os arquivos lado a lado
!paste frutas1.txt cores1.txt

# Remover os arquivos de exemplo
!rm frutas1.txt cores1.txt
````

#### 19. `sed` (Stream Editor)

Editor de fluxo usado para filtrar e transformar texto. Geralmente usado para substituir texto em arquivos ou streams.

Sintaxe básica para substituição: `sed 's/padrao_antigo/padrao_novo/g' arquivo`
- `s`: comando de substituição.
- `/padrao_antigo/`: o texto ou regex a ser procurado.
- `/padrao_novo/`: o texto de substituição.
- `/g`: (global) substitui todas as ocorrências na linha, não apenas a primeira.

````python
# Criar um arquivo de exemplo
!echo -e "hello world\nthis is a test\nworld hello" > exemplo_sed.txt

# Substituir 'world' por 'universe'
!sed 's/world/universe/g' exemplo_sed.txt

# Substituir a primeira ocorrência de 'is' por 'was' em cada linha
!sed 's/is/was/' exemplo_sed.txt

# Deletar linhas que contêm a palavra 'test'
!sed '/test/d' exemplo_sed.txt

# Remover o arquivo de exemplo
!rm exemplo_sed.txt
````

#### 20. `awk`

Linguagem de script para processamento de texto. Excelente para extrair e manipular colunas de dados ou campos de arquivos formatados.

A sintaxe básica é `awk 'pattern { action }' arquivo`.
- `pattern`: Uma condição que, se verdadeira, executa a `action`.
- `action`: Um conjunto de comandos a serem executados (por padrão, imprime a linha inteira).

Variáveis especiais:
- `$0`: A linha inteira.
- `$1, $2, ...`: O primeiro, segundo, ... campo da linha (delimitado por espaço por padrão).
- `NR`: Número da linha atual.
- `NF`: Número de campos na linha atual.
- `FS`: Field Separator (separador de campo, padrão é espaço).

````python
# Criar um arquivo de exemplo com dados tabulados
!echo -e "Nome Idade Cidade\nAlice 30 NewYork\nBob 24 London\nCharlie 35 Paris" > dados.txt

# Exibir o arquivo
!cat dados.txt

# Imprimir a primeira e a terceira coluna
!awk '{print $1, $3}' dados.txt

# Imprimir linhas onde a idade (segunda coluna) é maior que 25
!awk '$2 > 25 {print $0}' dados.txt

# Imprimir o número da linha e a linha inteira
!awk '{print NR, $0}' dados.txt

# Remover o arquivo de exemplo
!rm dados.txt
````
